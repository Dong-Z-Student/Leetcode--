## 题目一 数组串联
题干：给你一个长度为 n 的整数数组 $nums$ 。请你构建一个长度为 $2n$ 的答案数组 $ans$ ，数组下标 从 0 开始计数 ，对于所有 $0 <= i < n$ 的 $i$ ，满足下述所有要求：$ans[i] == nums[i]、ans[i + n] == nums[i]$,具体而言，$ans$ 由两个 $nums$ 数组 串联 形成。返回数组 $ans$ 。
```c++
class Solution {
public:
    vector<int> getConcatenation(vector<int>& nums) {
        int size=nums.size();
        vector<int> result(nums);
        for (int i=0;i<size;++i){
            result.push_back(nums[i]);
        }
        return result;
    }
};
```

## 题目二 重新排列数组
题干：给你一个数组 $nums$ ，数组中有 $2n$ 个元素，按 $[x1,x2,...,xn,y1,y2,...,yn]$ 的格式排列。请你将数组按 $[x1,y1,x2,y2,...,xn,yn]$ 格式重新排列，返回重排后的数组。
```c++
class Solution {
public:
    vector<int> shuffle(vector<int>& nums, int n) {
        vector<int> result; 
        for(int i=0;i<n;++i){
            result.push_back(nums[i]);
            result.push_back(nums[n+i]);
        }
        return result;
    }
};
```

题干；给定一个二进制数组 $nums$ ， 计算其中最大连续 1 的个数。
```c++
class Solution {
public:
    int findMaxConsecutiveOnes(vector<int>& nums) {
        int result=0;
        int number=0;
        for(int i=0;i<nums.size();++i){
            if(nums[i]==1){
                number+=1;
            }
            else{
                result=max(result,number);
                number=0;
            }
        }
        result=max(result,number);
        return result;
    }
};
```

正确代码：
```c++
class Solution {
public:
    vector<int> findErrorNums(vector<int>& nums) {
        int size=nums.size();
        vector<int> result(2,0);
        vector<int> temp(size,0);
        for(int num:nums){
            temp[num-1]++;
        }
        for(int k=0;k<size;++k){
            if(temp[k]==2){
                result[0]=k+1;
            }
            else if(temp[k]==0){
                result[1]=k+1;
            }
        }
        return result;
    }
};
```

## 题目五 有多少小于当前数字的数字
题干：给你一个数组 $nums$，对于其中每个元素 $nums[i]$，请你统计数组中比它小的所有数字的数目。换而言之，对于每个 $nums[i]$ 你必须计算出有效的 $j$ 的数量，其中 $j$ 满足 $j != i 且 nums[j] < nums[i]$ 。以数组形式返回答案。
```c++
class Solution {
public:
    vector<int> smallerNumbersThanCurrent(vector<int>& nums) {
        int size=nums.size();
        vector<int> result(size,0);
        int number=0;
        for(int i=0;i<size;++i){
            for(int j=0;j<size;++j){
                if(nums[i]>nums[j]){
                    number+=1;
                }
            }
            result[i]=number;
            number=0;
        }
        return result;
    }
};
```
直接在for循环里面创建number，还能够省去最后赋0，以及创建result时可以不设定大小，后续push_back即可  

## 题目六(与题目四类似) 找到所有数组中消失的数字
题干：给你一个含 $n$ 个整数的数组 $nums$ ，其中 $nums[i]$ 在区间 $[1, n]$ 内。请你找出所有在 $[1, n]$ 范围内但没有出现在 $nums$ 中的数字，并以数组的形式返回结果。
```c++
class Solution {
public:
    vector<int> findDisappearedNumbers(vector<int>& nums) {
        int size=nums.size();
        vector<int> result;
        vector<int> temp(size,0);
        for(int num:nums){
            temp[num-1]++;
        }
        for(int k=0;k<size;++k){
            if(temp[k]==0){
                result.push_back(k+1);
            }
        }
        return result;
    }
};
```

## 题目七 加一
题干：给定一个表示大整数的整数数组$digits$，其中$ digits[i]$是整数的第$i$位数字。这些数字按从左到右，从最高位到最低位排列。这个大整数不包含任何前导 0。将大整数加 1，并返回结果的数字数组。
```c++
class Solution {
public:
    vector<int> plusOne(vector<int>& digits) {
        int carry = 1;  
        int n = digits.size();
        vector<int> result;
        for(int i = n-1; i >= 0; --i){
            int num = digits[i] + carry;
            result.push_back(num % 10);
            carry = num / 10;
        }
        if(carry){
            result.push_back(carry);
        }
        reverse(result.begin(), result.end());
        return result;
    }
};
```
如果不是9 → +1直接返回,如果是9 → 变0继续进位

题干：给定一个整数数组$arr$，如果它是有效的山脉数组就返回true，否则返回false。让我们回顾一下，如果$arr$满足下述条件，那么它是一个山脉数组：
- $arr.length >= 3$
- 在$0 < i < arr.length - 1$条件下，存在$i$使得：
- $arr[0] < arr[1] < ... arr[i-1] < arr[i]$ 
- $arr[i] > arr[i+1] > ... > arr[arr.length - 1]$
```c++
class Solution {
public:
    bool validMountainArray(vector<int>& arr) {
        int n = arr.size();
        if(n < 3) return false;
        int i = 0;
        while(i+1 < n && arr[i] < arr[i+1]){
            i++;
        }
        if(i == 0 || i == n-1) return false;
        while(i+1 < n && arr[i] > arr[i+1]){
            i++;
        }
        return i == n-1;
    }
};
```

题干：给你一个数组 $target$ 和一个整数 $n$。给你一个空栈和两种操作："Push"：将一个整数加到栈顶。"Pop"：从栈顶删除一个整数。同时给定一个范围 $[1, n]$ 中的整数流。使用两个栈操作使栈中的数字（从底部到顶部）等于 $target$。你应该遵循以下规则：如果整数流不为空，从流中选取下一个整数并将其推送到栈顶。如果栈不为空，弹出栈顶的整数。如果，在任何时刻，栈中的元素（从底部到顶部）等于 $target$，则不要从流中读取新的整数，也不要对栈进行更多操作。请返回遵循上述规则构建 $target$ 所用的操作序列。如果存在多个合法答案，返回 任一 即可。
```c++
class Solution {
public:
    vector<string> buildArray(vector<int>& target, int n) {
        vector<string> result;
        int j=0;
        for(int i=0;i<n;++i){
            result.push_back("Push");
            if(i+1==target[j]){
                j++;
                if(j==target.size()){break;}
            }
            else{result.push_back("Pop");}
        }
        return result;
    }
};
```
每个元素都进栈，不符合再弹出来就好了，需要一个指针始终追踪是否到达目标栈的栈顶了

$((10 \times \frac{6}{(9 + 3) \times -11}) + 17) + 5$

## 题目三 函数独占时间
题干：有一个单线程CPU正在运行一个含有$n$道函数的程序。每道函数都有一个位于$0$和$n-1$之间的唯一标识符。函数调用存储在一个调用栈上当一个函数调用开始时，它的标识符将会推入栈中。而当一个函数调用结束时，它的标识符将会从栈中弹出。标识符位于栈顶的函数是当前正在执行的函数 。每当一个函数开始或者结束时，将会记录一条日志，包括函数标识符、是开始还是结束、以及相应的时间戳。给你一个由日志组成的列表$logs$ ，其中$logs[i]$表示第$i$条日志消息，该消息是一个按{function_id}:{"start" | "end"}:{timestamp}进行格式化的字符串。例如，0:start:3意味着标识符为0的函数调用在时间戳3的起始开始执行；而1:end:2意味着标识符为$1$的函数调用在时间戳$2$的末尾结束执行。注意，函数可以调用多次，可能存在递归调用 。函数的独占时间定义是在这个函数在程序所有函数调用中执行时间的总和，调用其他函数花费的时间不算该函数的独占时间。例如，如果一个函数被调用两次，一次调用执行$2$单位时间，另一次调用执行$1$单位时间，那么该函数的独占时间为$2 + 1 = 3$ 。以数组形式返回每个函数的独占时间 ，其中第$i$个下标对应的值表示标识符$i$的函数的独占时间。
### 自己的代码
```c++
class Solution {
public:
    struct Def{
        string def_sign;
        bool type;
        int time;
    };

    vector<string> splitString(const string&str,char delimiter){
        vector<string> result;
        stringstream ss(str);
        string token;
        while(getline(ss,token,delimiter)){
            result.push_back(token);
        }
        return result;
    }

    Def readLog(string log){
        Def def_obj;
        vector<string> parts=splitString(log,':');
        def_obj.def_sign=parts[0];
        if(parts[1]=="start"){
            def_obj.type=true;
        }else if(parts[1]=="end"){
            def_obj.type=false;
        }
        def_obj.time=stoi(parts[2]);
        return def_obj;
    }

    vector<int> exclusiveTime(int n, vector<string>& logs) {
        vector<int> result(n);
        vector<Def> st;
        for(int i=0;i<logs.size();++i){
            Def def_obj=readLog(logs[i]);
            if(def_obj.type){
                if(st.size()!=0){
                    result[stoi(st.back().def_sign)]+=(def_obj.time-st.back().time);
                }
                st.push_back(def_obj);
            }
            else{
                result[stoi(st.back().def_sign)]+=(def_obj.time-st.back().time+1);
                st.pop_back();
                if(!st.empty()){
                    st.back().time = def_obj.time + 1; // 关键：父函数从 end+1 继续
                }
            }
        }
        return result;
    }
};
```
父函数的开始时间是需要更新的，它下一次开始运行的时间不是上一次结束的时间，而是其调用函数结束的时间，这样才能保证全是独立运行时间
### 学习代码(高效且简洁)
```c++
class Solution {
public:
    vector<int> exclusiveTime(int n, vector<string>& logs) {
        // pair<id, start>
        stack<pair<int,int>> st;
        int ts = 0;
        vector<int> ans(n);
        for (auto& log : logs) {
            auto pos = log.find(':');
            int id = stoi(log.substr(0, pos));
            auto pos2 = log.find(':', pos+1);
            string op = log.substr(pos+1, pos2-pos-1);
            int time = stoi(log.substr(pos2+1));
            if (op == "start") {
                if (!st.empty()) {
                    auto last = st.top();
                    ans[last.first] += time - ts;
                }
                ts = time;
                st.push(make_pair(id, time));
            } 
            else {
                ans[id] += time + 1 - ts;
                ts = time + 1;
                st.pop();
            }
        }
        return ans;
    }
};
```
学习点1：  
find函数结合substr函数高效拆分字符串，利用find函数找到分割字段所在的位置，利用substr获取一段范围内的字符
学习点2：  
直接使用c++自带stack栈，避免vector数组作栈的冗余  
学习点3：  
使用pair二元组快速记录，无需自定义struct

## 题目一 商品折扣后的最终价格
题干：给你一个数组 $prices$ ，其中 $prices[i]$ 是商店里第 $i$ 件商品的价格。商店里正在进行促销活动，如果你要买第 $i$ 件商品，那么你可以得到与 $prices[j]$ 相等的折扣，其中 $j$ 是满足 $j > i 且 prices[j] <= prices[i]$ 的 最小下标 ，如果没有满足条件的 $j$ ，你将没有任何折扣。请你返回一个数组，数组中第 $i$ 个元素是折扣后你购买商品 $i$ 最终需要支付的价格。
```c++
class Solution {
public:
    vector<int> finalPrices(vector<int>& prices) {
        int n=prices.size();
        vector<int> result(n);
        for(int i=0;i<n;++i){
            result[i] = prices[i];
            for(int j=i+1;j<n;++j){
                if(prices[j]<=prices[i]){
                    result[i]=(prices[i]-prices[j]);break;
                }
            }
        }
        return result;
    }
};
```
暴力求解，两次for循环，时间复杂度$n^2$
```c++
class Solution {
public:
    vector<int> finalPrices(vector<int>& prices) {
        //pair<id,start>
        stack<pair<int,int>>st;
        vector<int> result(prices.size());
        for(int i=0;i<prices.size();++i){
            if(st.empty()){
                st.push(make_pair(i,prices[i]));
            }
            else{
                while(!st.empty()&&prices[i]<=st.top().second){
                    result[st.top().first]=st.top().second-prices[i];
                    st.pop();
                }
                st.push(make_pair(i,prices[i]));
            }
        }
        while(!st.empty()){
            result[st.top().first]=st.top().second;
            st.pop();
        }
        return result;
    }
};
```

```c++
class Solution {
public:
    vector<int> finalPrices(vector<int>& prices){
        stack<int> st;
        vector<int> result(prices);
        for(int i = 0; i < prices.size(); ++i){
            while(!st.empty() && prices[i] <= prices[st.top()]){
                result[st.top()] -= prices[i];
                st.pop();
            }
            st.push(i);
        }
        return result;
    }
};
```

## 题目二 每日温度
题干：给定一个整数数组 $temperatures$ ，表示每天的温度，返回一个数组 $answer$ ，其中 $answer[i]$ 是指对于第 $i$ 天，下一个更高温度出现在几天后。如果气温在这之后都不会升高，请在该位置用 $0$ 来代替。
```c++
class Solution {
public:
    vector<int> dailyTemperatures(vector<int>& temperatures) {
        int n=temperatures.size();
        vector<int> answer(n);
        stack<pair<int,int>> st;
        for(int i=0;i<n;++i){
            if(st.empty()){
                st.push(make_pair(i,temperatures[i]));
            }
            else{
                while(!st.empty()&&temperatures[i]>st.top().second){
                    answer[st.top().first]=i-st.top().first;
                    st.pop();
                }
                st.push(make_pair(i,temperatures[i]));
            }
        }
        while(!st.empty()){
            answer[st.top().first]=0;
            st.pop();
        }
        return answer;
    }
};
```
和上一题基本如出一辙没什么变化  
跳跃DP的思想：
```c++
class Solution {
public:
    vector<int> dailyTemperatures(vector<int>& temperatures) {
        int n = temperatures.size();
        vector<int> res(n,0);
        int max_tem = 0;
        for(int i = n-1;i>=0;i--){
            if(temperatures[i]>=max_tem){
                max_tem = temperatures[i];
                continue;
            }
            int day = 1;
            while(temperatures[i+day]<=temperatures[i]){
                day+=res[i+day];
            }  
//这个while语句就是实现跳跃的地方，具体解释为如果我发现A这个时间节点的温度大于后面B时间节点的温度，那么A、B间的时间我可以不用再比较了，同时我还知道B之后多久才会出现比B温度高的时间节点C，那么我直接跳转到C，将A与C的温度进行对比就行了
            res[i]=day;
        }
        return res;
    }
};
```

## 题目三 柱状图中最大的矩形 
题干：给定 n 个非负整数，用来表示柱状图中各个柱子的高度。每个柱子彼此相邻，且宽度为 1 。求在该柱状图中，能够勾勒出来的矩形的最大面积。
```c++
class Solution {
public:
    int largestRectangleArea(vector<int>& heights) {
        int max_area=0;
        int n=heights.size();
        stack<int> st;
        for(int i=0;i<=n;++i){
            int h = i==n ? 0 : heights[i];
            while(!st.empty()&&h<heights[st.top()]){
                int mid=st.top();
                st.pop();

                int left=st.empty()?-1:st.top();
                int width=i-left-1;

                max_area=max(heights[mid]*width,max_area);
            }
            st.push(i);
        }
        return max_area;
    }
};
```
本质就是利用单调栈找每个元素左右第一个更小  
维护一个递增栈（即栈内元素越来越大），如果出现一个比栈顶小的元素，证明要进行压栈弹栈的处理了，即如果当前柱子更矮说明之前的柱子右边界找到了，而这是一个递增栈左边元素一定更小，因此得到了特定柱子的可达的左右边界  

## 题目四 去除重复字母  
题干：给你一个字符串 s ，请你去除字符串中重复的字母，使得每个字母只出现一次。需保证 返回结果的字典序最小（要求不能打乱其他字符的相对位置）。
```c++
class Solution {
public:
    string removeDuplicateLetters(string s) {
        unordered_map<char,int>cnt;
        unordered_map<char,bool>inStack;
        string st;
        for(char c:s){
            cnt[c]++;
        }
        for(char c:s){
            cnt[c]--;
            if(inStack[c]){continue;}
            while(!st.empty()&&c<st.back()&&cnt[st.back()]>0){
                inStack[st.back()]=false;
                st.pop_back();
            }
            st.push_back(c);
            inStack[c]=true;
        }
        return st;
    }
};
```

```c++
class Solution {
public:
    string removeDuplicateLetters(string s) {
        vector<int> cnt(26, 0);       // 统计每个字符剩余出现次数
        vector<bool> inStack(26,false); // 标记字符是否已经在栈中
        string st;                    // 用 string 模拟栈
        // 统计字符出现次数
        for(char c : s){
            cnt[c - 'a']++;
        }
        for(char c : s){
            cnt[c - 'a']--;
            // 如果已经在栈里，直接跳过
            if(inStack[c - 'a']) continue;
            // 维护单调栈
            while(!st.empty() && st.back() > c && cnt[st.back() - 'a'] > 0){
                inStack[st.back() - 'a'] = false;
                st.pop_back();
            }
            st.push_back(c);
            inStack[c - 'a'] = true;
        }
        return st;
    }
};
```  

## 单调栈总结
单调栈是一种用于**在线性时间内寻找数组中某个元素左侧或右侧最近更大（或更小）元素**的算法技巧。许多题目本质上都可以转化为寻找“第一个更大元素”“第一个更小元素”或“最近更大/更小元素”的问题，例如寻找右侧第一个更高温度、寻找区间边界或计算最大面积等。  
单调栈的核心思想是**维护一个保持单调性的栈结构**，栈中通常**存储数组元素的索引**。根据问题需求，栈可以保持 单调递增（栈中元素从小到大）或 单调递减（栈中元素从大到小）。当遍历数组时，如果**当前元素破坏了栈的单调性，就不断弹出栈顶元素，并在弹出时计算该元素的答案**。例如，在寻找“右侧第一个更大元素”的问题中，如果当前元素大于栈顶元素，就说明当前元素正是栈顶元素的答案，因此可以在弹栈时记录结果。随后再将当前元素入栈，等待未来元素决定它的答案。  
首先判断题目是否涉及寻找某个元素左右最近的更大或更小元素；其次确定遍历方向，如果是寻找右侧元素通常从左到右遍历，如果是寻找左侧元素则从右到左遍历；然后确定栈的单调性，如果要寻找更大元素通常使用递减栈，如果要寻找更小元素通常使用递增栈；最后根据条件编写循环，在当前元素与栈顶元素满足条件时弹栈并计算答案。  
单调栈不仅可以用于寻找“下一个更大/更小元素”，还经常用于解决区间问题和面积问题。例如在**柱状图最大矩形**问题中，通过单调栈可以快速找到每个柱子左右第一个更矮的柱子，从而确定该柱子作为最矮高度时能扩展的最大宽度；在**接雨水**问题中，单调栈用于寻找形成凹槽结构的左右边界；在**移除K位数字**这样的贪心问题中，单调栈则用于维护一个最优序列。这些题目虽然形式不同，但本质都是**利用单调栈维护某种顺序并在条件触发时进行批量计算**。从更深层的角度看，单调栈的本质可以理解为一种**延迟计算**机制。当某个元素的答案暂时无法确定时，就先将它压入栈中；当未来出现能够确定答案的元素时，再一次性弹出并计算结果。  

# 队列
## 题目一 无法吃午餐的学生数量
题干：学校的自助午餐提供圆形和方形的三明治，分别用数字 0 和 1 表示。所有学生站在一个队列里，每个学生要么喜欢圆形的要么喜欢方形的。
餐厅里三明治的数量与学生的数量相同。所有三明治都放在一个栈里，每一轮：
1. 如果队列最前面的学生喜欢栈顶的三明治，那么会拿走它并离开队列。  
2. 否则，这名学生会放弃这个三明治 并回到队列的尾部。

这个过程会一直持续到队列里所有学生都不喜欢栈顶的三明治为止。给你两个整数数组students和sandwiches，其中sandwiches[i]是栈里面第i个三明治的类型（i = 0是栈的顶部），students[j]是初始队列里第j名学生对三明治的喜好（j = 0是队列的最开始位置）。请你返回无法吃午餐的学生数量。
### 自己代码
```c++
class Solution {
public:
    int countStudents(vector<int>& students, vector<int>& sandwiches) {
        int n=sandwiches.size();
        while(n!=0){
            if(students[0]==sandwiches[0]){
                sandwiches.erase(sandwiches.begin());
                students.erase(students.begin());
                n--;
            }
            else{
                int a=students[0];
                students.erase(students.begin());
                students.push_back(a);
            }
        }
        return students.size();
    }
};
```
存在很多问题，首先这个while循环是很可能死循环的，原因在于我错误的认为三明治一定会拿完，例如如果所有学生都是1，三明治都是0，三明治永远不可能拿完
### 改正后的代码
```c++
class Solution {
public:
    int countStudents(vector<int>& students, vector<int>& sandwiches) {
        queue<int> q;
        for(int s:students){q.push(s);}
        int index=0;
        int count=0;
        while(!q.empty()&&count<q.size()){
            if(q.front()==sandwiches[index]){
                index++;
                q.pop();
                count=0;
            }
            else{
                q.push(q.front());
                q.pop();
                count++;
            }
        }
        return q.size();
    }
};
```
用index记录栈顶，index的移动表示弹栈操作，当count==q.size()时表面剩余学生队列中的所有人都不吃栈顶的这个三明治，所以就可以结束了，剩余人数也就是吃不上三明治的人数
### 最优代码
```c++
class Solution {
public:
    int countStudents(vector<int>& students, vector<int>& sandwiches) {
        int count[2] = {0};
        for(int s : students)
            count[s]++;
        for(int s : sandwiches) {
            if(count[s] == 0)
                break;
            count[s]--;
        }
        return count[0] + count[1];
    }
};
```
这段代码不在乎学生的顺序，只要还有学生吃栈顶的三明治则证明还可以继续进行下去，如果没有学生吃栈顶的三明治了，那么也就可以停止了开始计数了  

## 题目二 买票需要的时间
题干：有 n 个人前来排队买票，其中第0人站在队伍最前方，第(n - 1)人站在队伍最后方。给你一个下标从0开始的整数数组tickets，数组长度为n，其中第i人想要购买的票数为tickets[i]。每个人买票都需要用掉恰好1秒 。一个人一次只能买一张票 ，如果需要购买更多票，他必须走到队尾重新排队（瞬间发生，不计时间）如果一个人没有剩下需要买的票，那他将会离开队伍。返回位于位置k（下标从0开始）的人完成买票需要的时间（以秒为单位）。
### 正常利用队列进行解决
```c++
class Solution {
public:
    int timeRequiredToBuy(vector<int>& tickets, int k) {
        int n=tickets.size();
        queue<int> q;
        vector<int>time(n);
        for(int i=0;i<n;++i){q.push(i);}
        while(!q.empty()){
            for(int i=0;i<n;++i){
                time[i]++;
            }
            if(tickets[q.front()]-1!=0){
                tickets[q.front()]--;
                q.push(q.front());
            }
            else{
                if(q.front()==k)break;
            }
            q.pop();
        }
        return time[k];
    }
};
```
### 队列模拟->数学化简
```c++
class Solution {
public:
    int timeRequiredToBuy(vector<int>& tickets, int k) {

        int ans = 0;

        for(int i = 0; i < tickets.size(); i++) {

            if(i <= k)
                ans += min(tickets[i], tickets[k]);
            else
                ans += min(tickets[i], tickets[k] - 1);
        }

        return ans;
    }
};
```
这方法的关键就是：我们只关心k的情况，可以得到这样一个结论，如何k买完票就截止的话，我们去观察所有人的情况，那么如果人排在k的前面，他们最多只能买他们需要的票数（当他们票数少于k的票数，可以在k之前买完）或者是k的票数（他们票数大于k的票数，但是因为k一旦买完就结束了，他们的次数不可能k的票数）；然后如果排在k的后面，他们最多只能买他们需要的票数（当他们票数少于k的票数，可以在k之前买完）或者是k的票数减一（他们票数大于k的票数，但是因为k一旦买完就结束了，同时他们又排在k后面，所以会比k少买一轮），综上我们最终统计他们买的次数就知道结果了

## 题目三 用栈实现队列
题干：请你仅使用两个栈实现先入先出队列。队列应当支持一般队列支持的所有操作（push、pop、peek、empty）：
实现 MyQueue 类：
- void push(int x) 将元素 x 推到队列的末尾
- int pop() 从队列的开头移除并返回元素
- int peek() 返回队列开头的元素
- boolean empty() 如果队列为空，返回 true ；否则，返回 false

说明：
- 你只能使用标准的栈操作 —— 也就是只有 push to top, peek/pop from top, size, 和 is empty 操作是合法的。
- 你所使用的语言也许不支持栈。你可以使用 list 或者 deque（双端队列）来模拟一个栈，只要是标准的栈操作即可。
```c++
class MyQueue {
public:
    stack<int>s1;
    stack<int>s2;

    MyQueue() {
        stack<int>s1;
        stack<int>s2;
    }
    
    void push(int x) {
        s1.push(x);
    }
    
    int pop() {
        while(!s1.empty()){
            s2.push(s1.top());
            s1.pop();
        }
        int a=s2.top();
        s2.pop();
        while(!s2.empty()){
            s1.push(s2.top());
            s2.pop();
        }
        return a;
    }
    
    int peek() {
        while(!s1.empty()){
            s2.push(s1.top());
            s1.pop();
        }
        int a=s2.top();
        while(!s2.empty()){
            s1.push(s2.top());
            s2.pop();
        }
        return a;
    }
    
    bool empty() {
        return s1.empty()&&s2.empty();
    }
};

/**
 * Your MyQueue object will be instantiated and called as such:
 * MyQueue* obj = new MyQueue();
 * obj->push(x);
 * int param_2 = obj->pop();
 * int param_3 = obj->peek();
 * bool param_4 = obj->empty();
 */
```

## 题目一 最后一块石头的重量
题干：有一堆石头，每块石头的重量都是正整数。
每一回合，从中选出两块 最重的 石头，然后将它们一起粉碎。假设石头的重量分别为 x 和 y，且 x <= y。那么粉碎的可能结果如下：
- 如果 x == y，那么两块石头都会被完全粉碎；
- 如果 x != y，那么重量为 x 的石头将会完全粉碎，而重量为 y 的石头新重量为 y-x。

最后，最多只会剩下一块石头。返回此石头的重量。如果没有石头剩下，就返回 0。
```c++
class Solution {
public:
    int lastStoneWeight(vector<int>& stones) {
        priority_queue<int>pq;
        int n=stones.size();
        for(int i=0;i<n;++i){
            pq.push(stones[i]);
        }
        int max1;
        int max2;
        while(pq.size()>=2){
            max1=pq.top();
            pq.pop();
            max2=pq.top();
            pq.pop();
            int n=max1-max2;
            if(n!=0){
                pq.push(n);
            }
        }
        return pq.empty()?0:pq.top();
    }
};
```
这种题目明显的标志就是：一直取最大两个/取最小两个，都可以采用最大堆/最小堆解决
### 用桶排序做
```c++
class Solution {
public:
    int lastStoneWeight(vector<int>& stones) {
        const int MAXW = 1000;
        vector<int> bucket(MAXW + 1, 0);
        // 统计重量出现次数
        for (int s : stones) {
            bucket[s]++;
        }
        int i = MAXW;
        while (i > 0) {
            if (bucket[i] == 0) {
                i--;
                continue;
            }
            // 偶数个直接消掉
            if (bucket[i] % 2 == 0) {
                bucket[i] = 0;
                i--;
                continue;
            }
            // 剩一个，需要找下一个
            int j = i - 1;
            while (j > 0 && bucket[j] == 0) {
                j--;
            }
            if (j == 0) return i;
            bucket[i]--;
            bucket[j]--;
            int diff = i - j;
            bucket[diff]++;
        }
        return 0;
    }
};
```
石头重量范围很小，最多只有 1000 种可能值，因此可以用计数桶来记录每种重量出现的次数。我们不需要真的存所有石头，只需要记录：重量 i 的石头有多少个,从最大重量往下处理。两种情况：
- 某重量出现偶数次，直接全部消失
- 某重量剩 1 个，找到下一个最大重量和它碰撞

当题目满足“值域小 + 需要统计频率 + 可以按数值顺序遍历”时，应优先考虑桶排序或计数排序，因为可以把时间复杂度从 O(n log n) 降到 O(n)，把数据按照数值范围分到多个“桶”中，然后对每个桶内部排序，最后按顺序把所有桶里的数据依次取出。本质上是一种 “空间换时间” 的思想。

## 题目二 查找和最小的K对数字
题干：给定两个以非递减顺序排列的整数数组nums1和nums2,以及一个整数k。定义一对值(u,v)，其中第一个元素来自nums1，第二个元素来自nums2。请找到和最小的k个数对(u1,v1),(u2,v2) ... (uk,vk) 。
### 粗暴解
```c++
class Solution {
public:
    vector<vector<int>> kSmallestPairs(vector<int>& nums1, vector<int>& nums2, int k) {
        vector<vector<int>> result;
        priority_queue<
            pair<int,vector<int>>,
            vector<pair<int,vector<int>>>,
            greater<pair<int, vector<int>>>
        > pq;
        int n1=nums1.size();
        int n2=nums2.size();
        for(int i=0;i<n1;++i){
            for(int j=0;j<n2;++j){
                pair<int,vector<int>> p={nums1[i]+nums2[j],{nums1[i],nums2[j]}};//注意即使是数组也要用{}构造
                pq.push(p);
            }
        }
        for(int i=0;i<k&&!pq.empty();++i){
            result.push_back(pq.top().second);
            pq.pop();
        }
        return result;
    }
};
```
这是一个很呆的硬解方法，两个for循环提取所有数字对，然后利用最小堆做排序，这种方法在初始数组很小的时候可以用，但是当数组很大时，会爆存储，会爆时间
### 优化解
思路：首先这两个数组都是有序数组，所有这两个数组两两结合构成的数字对矩阵是“单调矩阵”，即从左到右，从上到下是单调增的，那么既然这样，先存入（i，0），在这其中获取最小的一个，只有当（i，0）被挑选到了其后面的（i，1）...（i，n）才需要进入堆中（因为他们肯定大于（i，0），若是（i，0）都没被挑走，他们也没有机会），其余的类推这个思路，只有（i，j）被挑走（i，j+1）才进入堆中
```c++
class Solution {
public:
    vector<vector<int>> kSmallestPairs(vector<int>& nums1, vector<int>& nums2, int k) {
        vector<vector<int>> result;
        using T=tuple<int,int,int>;
        priority_queue<T,vector<T>,greater<T>>pq;
        for(int i=0;i<nums1.size();++i){
            pq.push({nums1[i]+nums2[0],i,0});
        }
        while(k-->0&&!pq.empty()){
            auto[sum,i,j]=pq.top();
            pq.pop();
            result.push_back({nums1[i],nums2[j]});
            if(j+1<nums2.size()){
                pq.push({nums1[i]+nums2[j+1],i,j+1});
            }
        }
        return result;
    }
};
```

## 题目三 多次求和构造目标数组
题干：给你一个整数数组 target 。一开始，你有一个数组 A ，它的所有元素均为 1 ，你可以执行以下操作：
- 令 x 为你数组里所有元素的和
- 选择满足 0 <= i < target.size 的任意下标 i ，并让 A 数组里下标为 i 处的值为 x 。
- 你可以重复该过程任意次

如果能从 A 开始构造出目标数组 target ，请你返回 True ，否则返回 False 。
### 错误代码（试图正向模拟构造过程，但是这种方法行不通）
```c++
class Solution {
public:
    bool isPossible(vector<int>& target) {
        int n=target.size();
        vector<int> temp(n,1);
        using T=pair<int,int>;//num,index
        priority_queue<T,vector<T>,greater<T>> pq;
        for(int i=0;i<n;++i){
            pq.push({target[i],i});
        }
        while(!pq.empty()){
            int sum=0;
            for(int x:temp){sum+=x;}
            if(sum==pq.top().first){
                temp[pq.top().second]=sum;
                pq.pop();
            }else{
                return false;
            }
        }
        return true;
    }
};
```
错误的用例为：[8,5],这个其实是可以构造出来的，但是在我当前代码中，第一次1+1=2！=5是程序就会判断无法构造出目标数组了  
这版代码的核心判断是“当前和是否等于某个目标值”，以及当存在不等值时就无法构造，但是问题是：就算等于，也未必说明走在正确路径上，就算不等于，也未必说明无解，所以这个判断本身就不是有效判据。
### 正确代码（反向推理构造过程）
思路：核心前提是最大值一定是最后构造出来的，那么我们再来想一想最大值是如何得来的呢，是对上一个状态的数组求和得到的，而求和得到的这个当前的最大值只会改变唯一一个位置的值，所有我们能够做差求得这个被改变位置在上一个状态时的值，那么上一个状态就能够被推导出来了，那么这个逻辑成立后，我们只需要依次递推回去即可了，看最终的结果是否符合[1,...,1]的格式
```c++
class Solution {
public:
    bool isPossible(vector<int>& target) {
        int n=target.size();
        priority_queue<int> pq;
        long long sum=0;
        for(int x:target){
            sum+=x;
            pq.push(x);
        }
        while(true){
            int max=pq.top();
            pq.pop();

            long long rest=sum-max;

            if(max==1||rest==1)return true;
            if(rest==0||max<=rest)return false;

            int prev=max%rest;//这种写法是一个加速写法，能够直接跳过多步，原始版本是max-rest
            if(prev==0)return false;

            pq.push(prev);
            sum=rest+prev;
        }
    }
};
```
关键是搞清楚怎么判断代码该结束了：
- 第一步：先看是否已经成功&nbsp;&nbsp;if (max == 1 || rest == 1) return true;
  - 最大值都变成 1 了，说明全 1，成功
  - 其余元素和为 1 时，也一定能退回全 1，成功
- 第二步：再看是否根本不可能有上一步&nbsp;&nbsp;if (rest == 0 || max <= rest) return false;
  - rest == 0：没有其他元素支撑，不可能
  - max <= rest：反推出的前一个值不为正，非法（prev=max-rest,如果这个条件成立，则prev<=0,这是非法的）
- 第三步：尝试还原上一步&nbsp;&nbsp;prev = max % rest;
  - 这一步是在高效地求“上一步的值”。
- 第四步：看还原结果是否合法&nbsp;&nbsp;if (prev == 0) return false;
  - 因为上一步数组元素必须是正整数，不能是 0。

## 题目一 检测大写字母
题干：我们定义，在以下情况时，单词的大写用法是正确的：
- 全部字母都是大写，比如 "USA" 。
- 所有字母都不是大写，比如 "leetcode" 。
- 只有首字母大写， 比如 "Google" 。

给你一个字符串 word 。如果大写用法正确，返回 true ；否则，返回 false 。
### 简单实现
```c++
class Solution {
public:
    bool detectCapitalUse(string word) {
        if(word.size()==1){return true;}
        if(word[0]>='A'&&word[0]<='Z'&&word[1]>='A'&&word[1]<='Z'){
            for(int i=2;i<word.size();++i){
                if(word[i]>='a'&&word[i]<='z'){
                    return false;
                }
            }
        }
        else if(word[0]>='A'&&word[0]<='Z'){
            for(int i=1;i<word.size();++i){
                if(word[i]>='A'&&word[i]<='Z'){
                    return false;
                }
            }
        }
        else{
            for(char x:word){
                if(x>='A'&&x<='Z'){
                    return false;
                }
            }
        }
        return true;
    }
};
```
学习点就是如何判断大小写字母：判断是否大写（word[0]>='A'&&word[0]<='Z'），判断是否小写（word[i]>='a'&&word[i]<='z'）  
### 优化实现
```c++
class Solution {
public:
    bool detectCapitalUse(string word) {
        if (word.size() <= 1) return true;

        bool allUpper = (word[0] >= 'A' && word[0] <= 'Z') &&
                        (word[1] >= 'A' && word[1] <= 'Z');

        for (int i = 1; i < word.size(); ++i) {
            if (allUpper) {
                if (!(word[i] >= 'A' && word[i] <= 'Z')) return false;
            } else {
                if (!(word[i] >= 'a' && word[i] <= 'z')) return false;
            }
        }
        return true;
    }
};
```
先根据前两个字母的状态判断后面的字母应该处在什么状态，相较于之前那个版本能够减少重复判断

## 题目二 密钥格式化
题干：给定一个许可密钥字符串 s，仅由字母、数字字符和破折号组成。字符串由 n 个破折号分成 n + 1 组。你也会得到一个整数 k 。  
我们想要重新格式化字符串 s，使每一组包含 k 个字符，除了第一组，它可以比 k 短，但仍然必须包含至少一个字符。此外，两组之间必须插入破折号，并且应该将所有小写字母转换为大写字母。  
返回重新格式化的许可密钥 。
```c++
class Solution {
public:
    string licenseKeyFormatting(string s, int k) {
        s.erase(remove(s.begin(), s.end(), '-'), s.end());
        for(char &c:s){c=toupper(c);}
        for(int i=s.size()-k;i>0;i-=k){s.insert(i,1,'-');}
        return s;
    }
};
```
代码从后往前进行插入，因为后面的长度固定是k不需要多余处理，很简单，插到最后前面剩多少就是多少了，如果从前面往后插入的话，还需要计算前面需要预留多少字符用于插入  
这个是从后往前插入，其实还可以从后往前构造，如何再逆转一下就好了，而且push操作比insert操作耗时更短

## 题目三 隐藏个人信息
题干：给你一条个人信息字符串 s ，可能表示一个邮箱地址 ，也可能表示一串电话号码 。返回按如下规则 隐藏个人信息后的结果：
电子邮件地址：
一个电子邮件地址由以下部分组成：
- 一个 名字 ，由大小写英文字母组成，后面跟着
- 一个 '@' 字符，后面跟着
- 一个 域名 ，由大小写英文字母和一个位于中间的 '.' 字符组成。'.' 不会是域名的第一个或者最后一个字符。

要想隐藏电子邮件地址中的个人信息：
- 名字 和 域名 部分的大写英文字母应当转换成小写英文字母。
- 名字 中间的字母（即，除第一个和最后一个字母外）必须用 5 个 "*****" 替换。

电话号码：
一个电话号码应当按下述格式组成：
- 电话号码可以由 10-13 位数字组成
- 后 10 位构成 本地号码
- 前面剩下的 0-3 位，构成 国家代码
- 利用 {'+', '-', '(', ')', ' '} 这些 分隔字符 按某种形式对上述数字进行分隔

要想隐藏电话号码中的个人信息：
- 移除所有 分隔字符
- 隐藏个人信息后的电话号码应该遵从这种格式：
  - "***-***-XXXX" 如果国家代码为 0 位数字
  - "+*-***-***-XXXX" 如果国家代码为 1 位数字
  - "+**-***-***-XXXX" 如果国家代码为 2 位数字
  - "+***-***-***-XXXX" 如果国家代码为 3 位数字
- "XXXX" 是最后 4 位 本地号码
```c++
class Solution {
public:
    string maskPII(string s) {
        char type='t';
        string ans;
        for(char &c : s){
            if(c == '@'){
                type = 'e';
            }
            c=tolower(c);
        }
        switch(type){
            case 'e':{
                int pos = s.find('@');
                ans += s[0];
                ans += "*****";
                ans += s[pos - 1];
                ans += s.substr(pos);
                break;
            }
            case 't':{
                string num;
                for (char c : s) {
                    if (isdigit(c)) {
                        num += c;
                    }
                }
                int countryLen = num.size() - 10;
                if (countryLen > 0) {
                    ans += "+";
                    ans += string(countryLen, '*');
                    ans += "-";
                }
                ans += "***-***-";
                ans += num.substr(num.size() - 4);
                break;
            }
        }
        return ans;
    }
};
```

## 题目四 重复的子字符串
题干：给定一个非空的字符串 s ，检查是否可以通过由它的一个子串重复多次构成。
```c++
class Solution {
public:
    bool repeatedSubstringPattern(string s) {
        int n=s.size();
        for(int i=1;i<=n/2;++i){
            if (n % i != 0) continue;
            int j=1;
            bool result=true;
            for(int j=i;j<n;j+=i){
                if(s.substr(0,i)!=s.substr(j,i)){
                    result=false;
                    break;
                }
            }
            if(result){return true;}
        }
        return false;
    }
};
```

## 题目五 旋转字符串
题干：给定两个字符串, s和goal。如果在若干次旋转操作之后，s能变成goal，那么返回true。  
s的旋转操作就是将s最左边的字符移动到最右边。 
- 例如, 若s = 'abcde'，在旋转一次之后结果就是'bcdea'。
```c++
class Solution {
public:
    bool rotateString(string s, string goal) {
        int n1=goal.size();int n2=s.size();
        if(n1!=n2){return false;}
        string dualS=s;
        dualS+=s;
        for(int i=1;i<n2+1;++i){
            if(dualS.substr(i-1,n1)==goal){
                return true;
            }
        }
        return false;
    }
};
```

## 题目六 重复叠加字符串匹配
题干：给定两个字符串 a 和 b，寻找重复叠加字符串 a 的最小次数，使得字符串 b 成为叠加后的字符串 a 的子串，如果不存在则返回 -1。  
注意：字符串 "abc" 重复叠加 0 次是 ""，重复叠加 1 次是 "abc"，重复叠加 2 次是 "abcabc"。
```c++
class Solution {
public:
    int repeatedStringMatch(string a, string b) {
        string Splus="";
        int i=1;
        while(true){
            Splus+=a;
            if(Splus.find(b)!=string::npos){
                return i;
            }
            if(Splus.size()>b.size()+a.size()){
                return -1;
            }
            i++;
        }
    }
};
```

## 题目一 删除排序链表中的重复元素
题干：给定一个已排序的链表的头head，删除所有重复的元素，使每个元素只出现一次。返回已排序的链表。
```c++
/**
 * Definition for singly-linked list.
 * struct ListNode {
 *     int val;
 *     ListNode *next;
 *     ListNode() : val(0), next(nullptr) {}
 *     ListNode(int x) : val(x), next(nullptr) {}
 *     ListNode(int x, ListNode *next) : val(x), next(next) {}
 * };
 */
class Solution {
public:
    ListNode* deleteDuplicates(ListNode* head) {
        if (head == nullptr) return head;
        ListNode* prev=head;
        ListNode* current=prev->next;
        while(current!=nullptr){
            if(prev->val==current->val){
                prev->next=current->next;
                current=current->next;
            }else{
                prev=current;
                current=current->next;
            }
        }
        return head;
    }
};
```
这个写法写复杂了，只需要一个指针即可，改正方法在下面给出，双指针主要是提前熟悉一下写法&nbsp;&nbsp;**指针获取值时需要使用->不能使用.**
```c++
class Solution {
public:
    ListNode* deleteDuplicates(ListNode* head) {
        ListNode* cur = head;
        while (cur != nullptr && cur->next != nullptr) {
            if (cur->val == cur->next->val) {
                cur->next = cur->next->next;
            } else {
                cur = cur->next;
            }
        }
        return head;
    }
};
```

## 题目二 奇偶链表
题干：给定单链表的头节点 head ，将所有索引为奇数的节点和索引为偶数的节点分别分组，保持它们原有的相对顺序，然后把偶数索引节点分组连接到奇数索引节点分组之后，返回重新排序的链表。  
第一个节点的索引被认为是奇数 ， 第二个节点的索引为偶数 ，以此类推。  
请注意，偶数组和奇数组内部的相对顺序应该与输入时保持一致。  
你必须在 O(1) 的额外空间复杂度和 O(n) 的时间复杂度下解决这个问题。
```c++
/**
 * Definition for singly-linked list.
 * struct ListNode {
 *     int val;
 *     ListNode *next;
 *     ListNode() : val(0), next(nullptr) {}
 *     ListNode(int x) : val(x), next(nullptr) {}
 *     ListNode(int x, ListNode *next) : val(x), next(next) {}
 * };
 */
class Solution {
public:
    ListNode* oddEvenList(ListNode* head) {
        if (head == nullptr||head->next == nullptr) return head;
        ListNode*uneven=head;
        ListNode*even=head->next;
        ListNode*evenHead=head->next;
        while(even!=nullptr&&even->next!=nullptr){
            uneven->next=even->next;
            uneven=uneven->next;

            even->next=uneven->next;
            even=even->next;
        }
        uneven->next=evenHead;
        return head;
    }
};
```
链表类题目的关键就是要弄清楚链表指针如何修改的，只要能够理解这个其他都简单了，必须借助绘图进行理解

## 题目三 反转链表
题干：给你单链表的头节点 head ，请你反转链表，并返回反转后的链表。
```c++
/**
 * Definition for singly-linked list.
 * struct ListNode {
 *     int val;
 *     ListNode *next;
 *     ListNode() : val(0), next(nullptr) {}
 *     ListNode(int x) : val(x), next(nullptr) {}
 *     ListNode(int x, ListNode *next) : val(x), next(next) {}
 * };
 */
class Solution {
public:
    ListNode* reverseList(ListNode* head) {
        if (head == nullptr||head->next == nullptr) return head;
        ListNode*prev=head;
        ListNode*current=head->next;
        prev->next=nullptr;//这种写法下切记需要断开头才行，不然会形成环路
        while(current!=nullptr){
            ListNode*temp=current->next;
            current->next=prev;
            prev=current;
            current=temp;
        }
        head=prev;
        return head;
    }
};
```
### 无需断开的写法
```c++
class Solution {
public:
    ListNode* reverseList(ListNode* head) {
        ListNode* prev = nullptr;
        ListNode* current = head;

        while (current != nullptr) {
            ListNode* temp = current->next;
            current->next = prev;
            prev = current;
            current = temp;
        }

        return prev;
    }
};
```

# 哈希表
## 题目一
题干：给定一个整数数组 nums 和一个整数目标值 target，请你在该数组中找出 和为目标值 target  的那 两个 整数，并返回它们的数组下标。  
你可以假设每种输入只会对应一个答案，并且你不能使用两次相同的元素。  
你可以按任意顺序返回答案。  
### 暴力求解
```c++
class Solution {
public:
    vector<int> twoSum(vector<int>& nums, int target) {
        vector<int>result;
        int n=nums.size();
        for(int i=0;i<n;++i){
            for(int j=i+1;j<n;++j){
                if(nums[i]+nums[j]==target){
                    result={i,j};
                    return result;
                }
            }
        }
        return result;
    }
};
```
### 哈希表求解
```c++
class Solution {
public:
    vector<int> twoSum(vector<int>& nums, int target) {
        unordered_map<int, int> mp;  // value -> index
        for (int i = 0; i < nums.size(); ++i) {
            int need = target - nums[i];
            if (mp.count(need)) {
                return {mp[need], i};
            }
            mp[nums[i]] = i;
        }
        return {};
    }
};
```
这个方法更快的原因在于查找目标元素时的循环被替换为了哈希表直接查找，时间复杂度由o(n)降到o(1)
做如下一些解释：
1. **字典 map 本身不一定就是哈希表**  

“字典”或者“映射”说的是这样一种数据结构需求：  
- 存一批 键值对 (key, value)
- 通过 key 快速找到对应的 value
- 支持插入、删除、查找

这叫 抽象数据类型。它只规定“要干什么”，不规定“内部一定怎么实现”。因此，字典可以有很多实现方式：
- 用 哈希表 实现
- 用 平衡二叉搜索树 实现
- 用 跳表 实现
- 用 Trie 实现（某些字符串场景）

2. **C++中，map和unordered_map不一样**
- std::map
  - 不是哈希表
  - 一般底层是红黑树
  - 键是**有序**的
  - 查找、插入、删除通常是O(log n)
  - 这种结构下快速查找利用的是树的有序性：
    - 当前键比目标大，去左子树
    - 当前键比目标小，去右子树
- std::unordered_map
  - 是哈希表
  - 键是**无序**的
  - 查找、插入、删除平均是O(1)，最坏可能O(n)
  - 这种结构下查找利用的是哈希的离散性：
    - 对输入的 key 做哈希计算
    - 得到一个位置或桶编号
    - 到那个桶里找对应键
    - 如果有冲突，再按冲突处理方式继续找

## 题目二 随机链表的复制
题干：给你一个长度为 n 的链表，每个节点包含一个额外增加的随机指针 random ，该指针可以指向链表中的任何节点或空节点。  
构造这个链表的 深拷贝。 深拷贝应该正好由 n 个 全新 节点组成，其中每个新节点的值都设为其对应的原节点的值。新节点的 next 指针和 random 指针也都应指向复制链表中的新节点，并使原链表和复制链表中的这些指针能够表示相同的链表状态。复制链表中的指针都不应指向原链表中的节点。  
例如，如果原链表中有 X 和 Y 两个节点，其中 X.random --> Y 。那么在复制链表中对应的两个节点 x 和 y ，同样有 x.random --> y。  
返回复制链表的头节点。  
用一个由 n 个节点组成的链表来表示输入/输出中的链表。每个节点用一个 [val, random_index] 表示：  
- val：一个表示 Node.val 的整数。
- random_index：随机指针指向的节点索引（范围从 0 到 n-1）；如果不指向任何节点，则为 null 。

你的代码只接受原链表的头节点 head 作为传入参数。
```c++
/*
// Definition for a Node.
class Node {
public:
    int val;
    Node* next;
    Node* random;
    
    Node(int _val) {
        val = _val;
        next = NULL;
        random = NULL;
    }
};
*/
class Solution {
public:
    Node* copyRandomList(Node* head) {
        if(head==nullptr){return nullptr;}
        unordered_map<Node*,Node*>mp;
        Node* cur=head;
        while(cur!=nullptr){
            mp[cur]=new Node(cur->val);
            cur=cur->next;
        }
        cur=head;
        while(cur!=nullptr){
            mp[cur]->next=mp[cur->next];
            mp[cur]->random=mp[cur->random];
            cur=cur->next;
        }
        return mp[head];
    }
};
```
哈希表是新节点和旧节点之间的映射关系，**新=mp[旧]**,新节点利用旧节点的值构建出来（第一轮循环），新节点需要连接到新节点才对，连接方向与旧节点之间连接方向相同，所以我们需要通过旧节点的连接关系得到连接方向（即需要连接的节点），同时通过旧节点映射到新节点

## 题目三 缺失的第一个正数
题干：给你一个未排序的整数数组 nums ，请你找出其中没有出现的最小的正整数。  
请你实现时间复杂度为 O(n) 并且只使用常数级别额外空间的解决方案。
```c++
class Solution {
public:
    int firstMissingPositive(vector<int>& nums) {
        unordered_map<int,int>mp;
        int Max=0;
        for(int i=0;i<nums.size();++i){
            if(nums[i]>0){
                mp[nums[i]]=i;
            }
            Max=max(Max,nums[i]);
        }
        for(int i=1;i<=Max;++i){
            if(!mp.count(i)){
                return i;
            }
        }
        return Max+1;
    }
};
```
这个代码虽然时间复杂度为o(n)，但是额外空间复杂度为o(n)，不是要求的o(1)
### 改进算法
直接在原数组上进行操作，这样产生的额外空间就降低到了要求的o(1),本质上还是在做哈希操作，但是在原数组上进行，原理为:  
只关心正整数，所以负数和0不考虑，找一个没出现的最小正整数，然后数组大小是n，可以这样考虑，假设所给数组的数全部不在1~n之间，那么最小的数就是1，例如：n=5，数组为6，7，8，9，10或n=5，数组为-1，-2，-4，-9，-7，如果给的数组包含1~n之间的数，那么将这些数放在他们本应该出现的位置这样可以判断到底哪些数没有出现过，然后找第一个没出现的也就是最小正整数了，所以通过交换操作将对应数字换到正确位置，同时那些不在1~n之间的数字留在原位占位即可，方便我们后续的比较操作
```c++
class Solution {
public:
    int firstMissingPositive(vector<int>& nums) {
        int n = nums.size();
        for (int i = 0; i < n; ++i) {
            while (nums[i] >= 1 &&
                   nums[i] <= n &&
                   nums[nums[i] - 1] != nums[i]) {
                swap(nums[i], nums[nums[i] - 1]);
            }//是while操作而不是if判断，交换操作只是把当前这个数换到它正确的位置去，即换进来的那个数可能也还是不正确，需要重复操作
            //for操作：哪些位置还需要检查
            //while操作：当前这个位置是否处理完成了
        }
        for (int i = 0; i < n; ++i) {
            if (nums[i] != i + 1) {
                return i + 1;
            }
        }
        return n + 1;
    }
};
```

## 题目一 找到最高海拔
题干：有一个自行车手打算进行一场公路骑行，这条路线总共由 n + 1 个不同海拔的点组成。自行车手从海拔为 0 的点 0 开始骑行。  
给你一个长度为 n 的整数数组 gain ，其中 gain[i] 是点 i 和点 i + 1 的 净海拔高度差（0 <= i < n）。请你返回 最高点的海拔 。
```c++
class Solution {
public:
    int largestAltitude(vector<int>& gain) {
        int n=gain.size();
        int maxHeight=0;
        int height=0;
        for(int i=0;i<n;++i){
            height+=gain[i];
            maxHeight=max(maxHeight,height);
        }
        return maxHeight;
    }
};
```
当前位置海拔等于前面所有海拔差之和

## 题目二 使数组和能被P整除
题干：给你一个正整数数组 nums，请你移除 最短 子数组（可以为 空），使得剩余元素的 和 能被 p 整除。 不允许 将整个数组都移除。  
请你返回你需要移除的最短子数组的长度，如果无法满足题目要求，返回 -1 。  
子数组 定义为原数组中连续的一组元素。
```c++
class Solution {
public:
    int minSubarray(vector<int>& nums, int p) {
        int n = nums.size();
        long long sum = 0;
        vector<long long> preSum(n);
        for (int i = 0; i < n; ++i) {
            sum += nums[i];
            preSum[i] = sum;
        }
        int x = sum % p;
        if (x == 0) return 0;
        unordered_map<int, int> mp;
        mp[0] = -1;   // 表示空前缀
        int ans = n;
        for (int j = 0; j < n; ++j) {
            int cur = preSum[j] % p;
            int need = (cur - x + p) % p;
            if (mp.count(need)) {
                ans = min(ans, j - mp[need]);
            }
            mp[cur] = j;
        }
        return ans == n ? -1 : ans;
    }
};
```
题目要求删除一个子数组后剩余元素的和能够整除目标值$ p $，那么在这个关系下，假设原数组和为$ SUM $，删除的子数组的和为$ sub $，删除后的和为$ SUM-sub $，最终满足$ (SUM-sub)\%p=0 $,整理得$ SUM\%p=sub\%p $，假设$ x=SUM\%p $，问题转化为找一个连续子数组，使得$ sub\%p=x $。引入前缀和$ pre $后，一个连续子数组的和$ sub=Pre[j]-pre[i-1] $，代入得$ (pre[j]-pre[i-1])\%p=x $，由同余定理得$ pre[j]-pre[i-1]≡x (mod p) $,移项得$ pre[j]-x≡pre[i-1] (mod p) $，即$ (pre[j]-x)\%p=pre[i-1]\%p $，根据上述公式知道我们的目标就是当我们在遍历右端点$ j $的时候，需要找到一个前缀使其满足上述表达式，再总结一下，设$ cur=pre[j]\%p $，我们的目标值$ need=(cur-x)\%p $,即找到一个$ i $使得$ pre[i]\%p=need $，数学层面总结即:寻找一个最短连续子数组，使其和对$ p $取模等于数组总和对$ p $的余数  
不能先存一遍再进行第二遍查找，因为这题里哈希表代表的不是“全局有哪些模值”，而是：在当前下标左边，已经出现过哪些前缀模值，以及它们的位置。对于当前 j，我们要做的是：在 j 左边所有前缀里，找一个位置 i，满足pre[i] % p == need这句话里，重点不是“有没有这个模值”，而是：
- 这个位置必须在左边
- 这个位置必须是历史位置
- 最好尽量靠近 j

所以哈希表必须随着遍历过程动态维护，而不能提前一次性建完。  

### 题目三 生成平衡数组的方案数
题干：给你一个整数数组 nums 。你需要选择 恰好 一个下标（下标从 0 开始）并删除对应的元素。请注意剩下元素的下标可能会因为删除操作而发生改变。  
比方说，如果 nums = [6,1,7,4,1] ，那么：
- 选择删除下标 1 ，剩下的数组为 nums = [6,7,4,1] 。
- 选择删除下标 2 ，剩下的数组为 nums = [6,1,4,1] 。
- 选择删除下标 4 ，剩下的数组为 nums = [6,1,7,4] 。
 
如果一个数组满足奇数下标元素的和与偶数下标元素的和相等，该数组就是一个平衡数组。  
请你返回删除操作后，剩下的数组 nums 是平衡数组的方案数。  
```c++
class Solution {
public:
    int waysToMakeFair(vector<int>& nums) {
        int n=nums.size();
        vector<int>evenPrefix(n,0);
        vector<int>oddPrefix(n,0);
        int evenSum=0;
        int oddSum=0;
        for(int i=0;i<n;++i){
            if(i%2==0){
                evenSum+=nums[i];
                evenPrefix[i]=evenSum;
                oddSum+=0;
                oddPrefix[i]=oddSum;
            }
            else{
                oddSum+=nums[i];
                oddPrefix[i]=oddSum;
                evenSum+=0;
                evenPrefix[i]=evenSum;
            }
        }
        int ans=0;
        for(int i=0;i<n;++i){
            int leftEven=(i>0?evenPrefix[i-1]:0);
            int leftOdd=(i>0?oddPrefix[i-1]:0);
            int rightEven=evenSum-evenPrefix[i];
            int rightOdd=oddSum-oddPrefix[i];
            int totalEven=leftEven+rightOdd;
            int totalOdd=leftOdd+rightEven;
            if(totalEven==totalOdd){
                ans+=1;
            }else{
                ans+=0;
            }
        }
        return ans;
    }
};
```
关键就是删除一个元素后，它前面的元素的奇偶性不会改变，但是它后面的元素奇偶会转变，利用这一特性就可以用奇偶和减去左侧的奇偶和得到右侧的奇偶和，再反转，就是需要的值了，那么删除过后的奇数和就是删除位置前的奇数和加删除位置后的原偶数和，偶数和同理，这里有一个关键的地方就是**存奇偶前缀和的时候不能只存奇数和偶数个数那么多个数，当某个位置不是奇数位或偶数位时这个地方应该+0来占位，这样后面才能正常获取到前缀和**，这两个数组的 索引 i 不是“第几个偶数项 / 第几个奇数项”，而是：原数组的下标范围 [0...i]，每个位置 i 都表示：截止到原数组位置 i，**偶数下标元素累计了多少/奇数下标元素累计了多少**

# 排序的前置知识
这里面刷题能用到的不多，但是每种排序的思想都需要去体会，这些思想有时候会提供很好的解题思路
## 常见的排序算法
### 比较排序
通过“比较两个元素大小”来决定先后顺序
包括：冒泡排序、选择排序、插入排序、希尔排序、快速排序、归并排序、堆排序
1. 冒泡排序
  - 核心思想

    不断比较相邻两个元素，如果顺序不对就交换。每一轮结束后，当前未排序部分中的最大值都会被“冒”到最后面。

  - 实现逻辑

    假设数组长度为 n：
      - 第 1 轮：把最大元素放到位置 n-1
      - 第 2 轮：把次大元素放到位置 n-2
      - ...
      - 第 n-1 轮：排序完成

    第 i 轮中，只需要比较到 n-1-i 即可，因为后面的元素已经有序。

  - 时空复杂度
    - 最好时间复杂度：O(n),数组本来就有序，第一轮发现没有交换，直接结束
    - 平均时间复杂度：O(n^2)
    - 最坏时间复杂度：O(n^2)
    - 空间复杂度：O(1)
    - 稳定排序,因为相等元素不会被强制交换相对顺序。

  - 代码示例
    ```c++
    #include <vector>
    using namespace std;
    void bubbleSort(vector<int>& nums) {
        int n = nums.size();
        for (int i = 0; i < n - 1; ++i) {
            bool swapped = false;
            for (int j = 0; j < n - 1 - i; ++j) {
                if (nums[j] > nums[j + 1]) {
                    swap(nums[j], nums[j + 1]);
                    swapped = true;
                }
            }
            if (!swapped) break;
        }
    }
    ```
2. 选择排序
  - 核心思想

    每一轮从未排序区间中选择最小元素，放到当前应放的位置。它不像冒泡那样不断交换相邻元素，而是“先找到最小值，再一次性放过去”。

  - 实现逻辑

    假设数组长度为 n：
    - 第 1 轮：在 [0, n-1] 中找最小值，放到 0
    - 第 2 轮：在 [1, n-1] 中找最小值，放到 1
    - ...
    - 第 n-1 轮：排序完成

    每一轮只需要记录最小值下标，最后交换一次。

  - 时空复杂度
    - 最好时间复杂度：O(n^2)
    - 平均时间复杂度：O(n^2)
    - 最坏时间复杂度：O(n^2)
    - 空间复杂度：O(1)
    - 不稳定排序

  - 代码示例
    ```c++
    #include <vector>
    using namespace std;
    void selectionSort(vector<int>& nums) {
        int n = nums.size();
        for (int i = 0; i < n - 1; ++i) {
            int minIndex = i;
            for (int j = i + 1; j < n; ++j) {
                if (nums[j] < nums[minIndex]) {
                    minIndex = j;
                }
            }
            if (minIndex != i) {
                swap(nums[i], nums[minIndex]);
            }
        }
    }
    ```
3. 插入排序
  - 核心思想

    插入排序的思想非常像打扑克牌：每次从未排序部分拿出一个元素，插入到前面已经有序的部分中的合适位置。

  - 实现逻辑
    将数组分为两部分：
    - 左边：已排序区间
    - 右边：未排序区间

    开始时，默认第一个元素自己就是有序的。
    然后从第二个元素开始：
    - 取出当前元素 key
    - 从前往后已经有序的部分中，找到插入位置
    - 为了插入，需要把所有比 key 大的元素向后移动一位
    - 最后把 key 放进去

  - 时空复杂度
    - 最好时间复杂度：O(n)
      - 原数组本身有序时，每次只比较一次
      - 平均时间复杂度：O(n^2)
      - 最坏时间复杂度：O(n^2)
      - 空间复杂度：O(1)
      - 稳定排序
  - 代码示例
    ```c++
    #include <vector>
    using namespace std;
    void insertionSort(vector<int>& nums) {
        int n = nums.size();
        for (int i = 1; i < n; ++i) {
            int key = nums[i];
            int j = i - 1;
            while (j >= 0 && nums[j] > key) {
                nums[j + 1] = nums[j];
                --j;
            }
            nums[j + 1] = key;
        }
    }
    ```
4. 希尔排序
  - 核心思想

    希尔排序可以看成是插入排序的改进版。  
    插入排序的问题在于：如果一个很小的元素在数组最后面，想把它挪到前面，要移动很多次  
    希尔排序的思路是：  
    先让较远距离的元素进行比较和交换，逐步缩小间隔，最后再做一次普通插入排序。  
    这样可以提前把元素大致调整到接近正确位置，从而减少最终插入排序的移动次数。  

  - 实现逻辑

    引入一个“步长” gap：
    - 初始设为 n/2
    - 每轮让所有间隔为 gap 的元素做“分组插入排序”
    - 然后不断缩小 gap
    - 最后当 gap = 1 时，就是一次普通插入排序

  - 时空复杂度
    希尔排序的复杂度和“增量序列”有关，不同写法不同。
    - 平均时间复杂度：通常认为约为 O(n^{1.3}) 到 O(n^2) 之间
    - 最坏时间复杂度：取决于 gap 设计，经典写法通常可视作 O(n^2)
    - 空间复杂度：O(1)
    - 不稳定排序
  - 代码示例
    ```c++
    #include <vector>
    using namespace std;
    void shellSort(vector<int>& nums) {
        int n = nums.size();
        for (int gap = n / 2; gap > 0; gap /= 2) {
            for (int i = gap; i < n; ++i) {
                int temp = nums[i];
                int j = i;
                while (j >= gap && nums[j - gap] > temp) {
                    nums[j] = nums[j - gap];
                    j -= gap;
                }
                nums[j] = temp;
            }
        }
    }
    ```
5. 快速排序
  - 核心思想

    快速排序的思想是分治，它的关键在于“划分”。：
    - 选一个基准值 pivot
    - 把数组分成两部分：
        - 小于等于 pivot 的放左边
        - 大于等于 pivot 的放右边
    - 然后递归排序左右两部分

  - 实现逻辑

    最常见的做法：
    - 取区间第一个元素作为基准
    - 定义两个指针 i 和 j
    - 从右往左找小于 pivot 的元素
    - 从左往右找大于 pivot 的元素
    - 如果 i < j，交换它们
    - 最终把 pivot 放到中间正确位置
    - 然后递归左右两部分
    
  - 时空复杂度
    - 最好时间复杂度：O(n log n)
    - 平均时间复杂度：O(n log n)
    - 最坏时间复杂度：O(n^2)
        - 比如数组本来有序，而你总是选第一个元素作 pivot
    - 空间复杂度：O(log n)
        - 主要是递归栈；最坏可到 O(n)
    - 不稳定排序
  - 代码示例
    ```c++
    //采用Hoare切分方法，即双指针从两端向中间进行扫描
    #include <vector>
    using namespace std;
    int partition(vector<int>& nums, int left, int right) {
        int pivot = nums[left];
        int i = left, j = right;
        while (i < j) {
            while (i < j && nums[j] >= pivot) --j;
            while (i < j && nums[i] <= pivot) ++i;
            if (i < j) swap(nums[i], nums[j]);
        }
        swap(nums[left], nums[i]);
        return i;
    }
    void quickSort(vector<int>& nums, int left, int right) {
        if (left >= right) return;
        int pivotIndex = partition(nums, left, right);
        quickSort(nums, left, pivotIndex - 1);
        quickSort(nums, pivotIndex + 1, right);
    }   
    quickSort(nums, 0, nums.size() - 1);
    ```
    ```c++
    //采用Lomuto切分方法，也是双指针，但是更简单，一个指针用于遍历，另一个指针用于记录小于pivot值的数组右边界即可
    int partition(vector<int>& nums,int l,int r){
        int pivot = nums[r];
        int i = l - 1;
        for(int j = l; j < r; j++){
            if(nums[j] < pivot){
                i++;
                swap(nums[i], nums[j]);
            }
        }
        swap(nums[i+1], nums[r]);
        return i + 1;
    }
    void quickSort(vector<int>& nums,int l,int r){
        if(l >= r) return;
        int p = partition(nums,l,r);
        quickSort(nums,l,p-1);
        quickSort(nums,p+1,r);
    }
    quickSort(nums,0,nums.size()-1);
    ```
    - 举例说明

    示例数组：[6,3,8,5,2,7]

    1. 初始状态

        数组:  [6, 3, 8, 5, 2, 7]  
        ↑              ↑  
        i              j  
        pivot = 6  

    2. j 从右向左找 < pivot

        j开始移动：7 >= 6  → j--  
        [6, 3, 8, 5, 2, 7]  
        ↑            ↑  
        i            j  
        继续：2 < 6  → 停止  
        [6, 3, 8, 5, 2, 7]  
        ↑            ↑  
        i            j  

    3. i 从左向右找 > pivot

        i开始移动：3 <= 6  → i++  
        [6, 3, 8, 5, 2, 7]  
            ↑        ↑  
            i        j  
        继续：8 > 6  → 停止  
        [6, 3, 8, 5, 2, 7]  
               ↑     ↑  
               i     j  

    4. 交换 i 和 j

        交换前[6, 3, 8, 5, 2, 7]  
        交换后[6, 3, 2, 5, 8, 7]  

    5. j 继续向左移动

        8 >= 6  → j--  
        [6, 3, 2, 5, 8, 7]  
               ↑  
               i  
                  ↑  
                  j  
        5 < 6  → 停止  

    6. i 继续向右移动

        2 <= 6 → i++  
        [6, 3, 2, 5, 8, 7]  
                  ↑  
                  i  
                  ↑  
                  j  
        此时i >= j，循环结束

    7. pivot归位

        交换pivot (6) 和 nums[j]  
        交换前[6, 3, 2, 5, 8, 7]  
        交换后[5, 3, 2, 6, 8, 7]  

    8. 分区完成

        [5, 3, 2] | 6 | [8, 7]  
        此时：
        - 6 左边全部小于 6
        - 6 右边全部大于 6
        - 6 已经在最终位置

    9. 递归继续

        继续排序：左边[5,3,2]、右边[8,7]
        递归结构：  
                [6,3,8,5,2,7]  
                       |  
                    pivot=6  
                       |  
          [5,3,2]      6      [8,7]  
             |                   |  
          pivot=5             pivot=8  
             |                   |  
           [2,3]               [7]  

    10. 最终结果

        [2,3,5,6,7,8]

    11. Partition 核心逻辑总结
        1. 右指针 j 找 < pivot
        2. 左指针 i 找 > pivot
        3. 交换 nums[i] 和 nums[j]
        4. 重复直到 i >= j
        5. 最后 pivot 与 nums[j] 交换

6. 归并排序
  - 核心思想

    归并排序也是分治思想：
    - 把数组不断二分
    - 递归排序左右两部分
    - 将两个有序数组合并成一个更大的有序数组

    它的重点不是“交换”，而是“合并”。

  - 实现逻辑

  假设区间为 [left, right]：
    - 如果区间长度为 1，天然有序
    - 否则取中点 mid
    - 递归排序 [left, mid] 和 [mid+1, right]
    - 然后使用双指针把两个有序区间合并到临时数组中
    - 再拷贝回原数组

  - 时空复杂度
    - 最好时间复杂度：O(n log n)
    - 平均时间复杂度：O(n log n)
    - 最坏时间复杂度：O(n log n)
    - 空间复杂度：O(n)
    - 稳定排序
  - 代码示例
    ```c++
    #include <vector>
    using namespace std;
    void merge(vector<int>& nums, int left, int mid, int right) {
        vector<int> temp;
        int i = left, j = mid + 1;
        while (i <= mid && j <= right) {
            if (nums[i] <= nums[j]) {
                temp.push_back(nums[i]);
                ++i;
            } else {
                temp.push_back(nums[j]);
                ++j;
            }
        }
        while (i <= mid) {
            temp.push_back(nums[i]);
            ++i;
        }
        while (j <= right) {
            temp.push_back(nums[j]);
            ++j;
        }
        for (int k = 0; k < (int)temp.size(); ++k) {
            nums[left + k] = temp[k];
        }
    }
    void mergeSort(vector<int>& nums, int left, int right) {
        if (left >= right) return;
        int mid = left + (right - left) / 2;
        mergeSort(nums, left, mid);
        mergeSort(nums, mid + 1, right);
        merge(nums, left, mid, right);
    }
    mergeSort(nums, 0, nums.size() - 1);
    ```
7. 堆排序
    - 核心思想

        堆排序利用的是堆这种数据结构。  
        如果要升序排序，就使用大根堆:  
        - 堆顶永远是最大元素
        - 每次把堆顶取出来，放到数组末尾
        - 然后对剩余部分重新调整为大根堆

        最终数组就有序了。

    - 实现逻辑

        堆排序分两步：
        - 第一步：建堆

        把原数组原地调整成一个大根堆。
        - 第二步：反复取堆顶

            - 堆顶元素和当前最后一个元素交换
            - 堆的有效长度减 1
            - 对新的堆顶向下调整

        这里最核心的操作是 heapify（下沉调整）。
    - 时空复杂度

        - 建堆：O(n)
        - 每次取堆顶并调整：O(log n)
        - 总时间复杂度：O(n log n)
        - 空间复杂度：O(1)
        - 不稳定排序

    - 代码示例
    ```c++
    #include <vector>
    using namespace std;
    void heapify(vector<int>& nums, int n, int i) {
        int largest = i;
        int left = 2 * i + 1;
        int right = 2 * i + 2;
        if (left < n && nums[left] > nums[largest]) {
            largest = left;
        }
        if (right < n && nums[right] > nums[largest]) {
            largest = right;
        }
        if (largest != i) {
            swap(nums[i], nums[largest]);
            heapify(nums, n, largest);
        }
    }
    void heapSort(vector<int>& nums) {
        int n = nums.size();
        // 建大根堆
        for (int i = n / 2 - 1; i >= 0; --i) {
            heapify(nums, n, i);
        }
        // 依次把堆顶放到末尾
        for (int i = n - 1; i > 0; --i) {
            swap(nums[0], nums[i]);
            heapify(nums, i, 0);
        }
    }
    ```
### 非比较排序
不完全依赖元素之间的大小比较，而是利用数据范围、位数、映射关系
包括：计数排序、桶排序、基数排序
1. 计数排序
    - 核心思想

    计数排序适用于：数据范围不大、且元素是整数的场景。它不比较元素大小，而是统计每个数字出现了多少次。
    例如：4、2、2、8、3、3、1，统计结果为：1 出现 1 次、2 出现 2 次、3 出现 2 次、4 出现 1 次、8 出现 1 次，然后按顺序把数字写回去即可。

    - 实现逻辑

        - 简单版逻辑
            - 找到最大值和最小值
            - 开一个计数数组 count
            - count[x] 表示数字 x 出现次数
            - 最后按数字从小到大，把它们按次数写回原数组

        - 稳定版逻辑  

            如果要求稳定，就不能只“写次数”，还要做前缀和，确定每个元素应该放在哪个位置。
    - 时空复杂度

        - 设数据范围大小为 k。  
            - 时间复杂度：O(n + k)
            - 空间复杂度：O(k)
        - 稳定性
            - 简单版：通常看作不强调稳定性
            - 前缀和版：可以稳定

        - 使用条件  

        计数排序很强，但限制也很明显：  
        - 必须适合整数
        - 数据范围不能太大

        比如：
        - n = 100
        - 元素值范围是 0 ~ 10^9

        这时就完全不适合计数排序。
    - 代码示例
    ```c++
    #include <vector>
    #include <algorithm>
    using namespace std;

    void countingSort(vector<int>& nums) {
        if (nums.empty()) return;
    
        int minVal = *min_element(nums.begin(), nums.end());
        int maxVal = *max_element(nums.begin(), nums.end());
    
        vector<int> count(maxVal - minVal + 1, 0);
    
        for (int x : nums) {
            count[x - minVal]++;
        }
    
        int index = 0;
        for (int i = 0; i < count.size(); ++i) {
            while (count[i] > 0) {
                nums[index++] = i + minVal;
                count[i]--;
            }
        }
    }
    ```
2. 桶排序
  - 核心思想

    桶排序的思想是：把数据分到若干个桶中，每个桶内部再分别排序，最后按桶顺序依次取出。
    它适合：
    - 数据分布比较均匀
    - 能够比较自然地映射到若干区间

    比如学生成绩 0~100，可以分成：0~9、10~19、20~29、...

  - 实现逻辑
    以整数为例：
    - 先求出最小值和最大值
    - 决定每个桶装多少范围，比如每个桶范围为 bucketSize
    - 根据元素值计算它属于哪个桶
    - 把元素放入对应桶中
    - 对每个桶内部单独排序
    - 依次收集每个桶的元素
  - 时空复杂度

    桶排序复杂度并不是固定的，它和数据分布关系很大。
    - 理想情况：数据均匀分布，每个桶元素很少：时间复杂度接近：O(n + k)，其中 k 是桶数。
    - 最坏情况：所有元素都落进同一个桶里，此时就退化成桶内排序的复杂度，比如 O(n log n) 或更差。
    - 空间复杂度：O(n + k)
    - 稳定性

    桶排序是否稳定，取决于：
    - 分桶过程是否保持顺序
    - 桶内排序是否稳定

    所以它可以稳定，也可以不稳定。
  - 代码示例
    ```c++
    #include <vector>
    #include <algorithm>
    using namespace std;
    void bucketSort(vector<int>& nums) {
        if (nums.empty()) return;
        int minVal = *min_element(nums.begin(), nums.end());
        int maxVal = *max_element(nums.begin(), nums.end());
        int bucketSize = 5;  // 每个桶覆盖的数值范围
        int bucketCount = (maxVal - minVal) / bucketSize + 1;
        vector<vector<int>> buckets(bucketCount);
        for (int x : nums) {
            int index = (x - minVal) / bucketSize;
            buckets[index].push_back(x);
        }
        int idx = 0;
        for (auto& bucket : buckets) {
            sort(bucket.begin(), bucket.end());
            for (int x : bucket) {
                nums[idx++] = x;
            }
        }
    }
    ```
3. 基数排序
  - 核心思想

    基数排序的思想是：按位排序。对整数来说，可以先按个位排，再按十位排，再按百位排……直到最高位。前提是：每一位的排序必须是稳定的。通常会借助“计数排序”来完成每一位的稳定排序。

  - 实现逻辑

    以十进制整数为例，从低位到高位排序（LSD，Least Significant Digit）：
    - 第 1 轮：按个位排序
    - 第 2 轮：按十位排序
    - 第 3 轮：按百位排序

  - 时空复杂度

    设：
    - n 是元素个数
    - d 是最大数的位数
    - k 是每一位可能取值范围（十进制下就是 10）

    则：
    - 时间复杂度：O(d(n + k))
    - 当 k 为常数时，通常写作 O(dn)
    - 空间复杂度：O(n + k)

    - 稳定排序因为每一位使用的是稳定排序。

    - 使用条件，基数排序适合：
        - 整数
        - 位数不太大
        - 可以按位拆解
  - 代码示例
    ```c++
    #include <vector>
    #include <algorithm>
    using namespace std;
    void countingSortByDigit(vector<int>& nums, int exp) {
        int n = nums.size();
        vector<int> output(n);
        vector<int> count(10, 0);
        for (int i = 0; i < n; ++i) {
            int digit = (nums[i] / exp) % 10;
            count[digit]++;
        }
        for (int i = 1; i < 10; ++i) {
            count[i] += count[i - 1];
        }
        for (int i = n - 1; i >= 0; --i) {
            int digit = (nums[i] / exp) % 10;
            output[count[digit] - 1] = nums[i];
            count[digit]--;
        }
        for (int i = 0; i < n; ++i) {
            nums[i] = output[i];
        }
    }
    void radixSort(vector<int>& nums) {
        if (nums.empty()) return;
        int maxVal = *max_element(nums.begin(), nums.end());
        for (int exp = 1; maxVal / exp > 0; exp *= 10) {
            countingSortByDigit(nums, exp);
        }
    }
    ```

1. “交换类”
    - 冒泡排序：相邻交换，把最大值冒到后面
    - 快速排序：通过 pivot 分区交换
2. “选择类”
    - 选择排序：每轮选最小

    - 快速排序：拆开再递归分区
5. “映射类”
    - 计数排序：统计次数
    - 桶排序：分桶处理
    - 基数排序：按位处理

## 题目一 最小绝对差
题干：给你个整数数组 arr，其中每个元素都 不相同。
请你找到所有具有最小绝对差的元素对，并且按升序的顺序返回。
每对元素对 [a,b] 如下：
- a , b 均为数组 arr 中的元素
- a < b
- b - a 等于 arr 中任意两个元素的最小绝对差
```c++
class Solution {
public:
    vector<vector<int>> minimumAbsDifference(vector<int>& arr) {
        vector<vector<int>>result;
        sort(arr.begin(),arr.end());
        int n=arr.size();
        int minAbs=abs(arr[0]-arr[1]);
        for(int i=0;i<n-1;++i){
            minAbs=min(abs(arr[i]-arr[i+1]),minAbs);
        }
        for(int i=0;i<n-1;++i){
            if(abs(arr[i]-arr[i+1])==minAbs){
                result.push_back({arr[i],arr[i+1]});;
            }
        }
        return result;
    }
};
```
c++中的sort函数用法：**sort(vector.begin(),vector.end())**

## 题目二 使数组元素相等的减少操作次数
题干：给你一个整数数组 nums ，你的目标是令 nums 中的所有元素相等。完成一次减少操作需要遵照下面的几个步骤：  
- 找出 nums 中的 最大 值。记这个值为 largest 并取其下标 i （下标从 0 开始计数）。如果有多个元素都是最大值，则取最小的 i 。
- 找出 nums 中的 下一个最大 值，这个值 严格小于 largest ，记为 nextLargest 。
- 将 nums[i] 减少到 nextLargest 。  

返回使 nums 中的所有元素相等的操作次数。
### 自己解
```c++
class Solution {
public:
    int reductionOperations(vector<int>& nums) {
        map<int,int>mp;
        int result=0;int level=0;
        for(int i=0;i<nums.size();++i){
            mp[nums[i]]+=1;
        }
        for(auto &p:mp){
            result+=p.second*level;
            level++;
        }
        return result;
    }
};
```
每个数出现的次数决定有几个这样的数要处理，数组里面唯一数的个数决定了每个数需要操作的次数，乘积求和即可
### 优化解（空间上更高效，不用新增存储空间）
```c++
class Solution {
public:
    int reductionOperations(vector<int>& nums) {
        sort(nums.begin(), nums.end());
        int result = 0;
        int cnt = 0;
        for (int i = 1; i < nums.size(); ++i) {
            if (nums[i] != nums[i - 1]) {
                cnt++;
            }
            result += cnt;
        }
        return result;
    }
};
```
排序之后每发送一次变化就意味着后面更大的数要多做一次“减少”操作了

## 题目三 合并区间
题干：以数组 intervals 表示若干个区间的集合，其中单个区间为 intervals[i] = [starti, endi] 。请你合并所有重叠的区间，并返回 一个不重叠的区间数组，该数组需恰好覆盖输入中的所有区间 。
```c++
class Solution {
public:
    vector<vector<int>> merge(vector<vector<int>>& intervals) {
        sort(intervals.begin(), intervals.end());
        int i = 0;
        while (i < intervals.size() - 1) {
            if (intervals[i][1] >= intervals[i + 1][0]) {
                intervals[i][1] = max(intervals[i][1], intervals[i + 1][1]);
                intervals.erase(intervals.begin() + i + 1);
            } else {
                i++;
            }
        }
        return intervals;
    }
};
```
要注意的就是有一个用例[[1,4],[2,3]]启发我这里得加一个max才行，不然最终的合并结果就是[1,3]

## 题目四 数组中的第K个最大元素
题干：给定整数数组 nums 和整数 k，请返回数组中第 k 个最大的元素。  
请注意，你需要找的是数组排序后的第 k 个最大的元素，而不是第 k 个不同的元素。  
你必须设计并实现时间复杂度为 O(n) 的算法解决此问题。  
### 堆排序算法
```c++
class Solution {
public:
    int findKthLargest(vector<int>& nums, int k) {
        priority_queue<int>pq;
        for(int n:nums){
            pq.push(n);
        }
        for(int i=0;i<k-1;++i){
            pq.pop();
        }
        return pq.top();
    }
};
```
但是堆排序的时间复杂度其实不是o(n)
### 快速排序->快速选择算法
确实如果要对一个数组进行全排序是没有能够将时间复杂度控制在o(n)的算法的，但是这道题由题意其实不需要进行全排序，只需要选择出目标元素即可，按照这个思路，目标元素在全排序结果中的位置是可确定的，即n-k，那么我们只需要在排序过程中“选择”出这个元素即可，那么分析各种排序算法的特性，能够在排序过程中就确定下来元素位置的算法就只有一种了，那就是快速排序算法，针对这个算法做优化即可，那么为什么这个算法优化后能够控制时间复杂度为o(n)呢，因为不需要递归了，这个过程中我们是能够控制到左边分区去继续查找还是去右边分区继续查找的。
```c++
class Solution {
public:
    int findKthLargest(vector<int>& nums, int k) {
        int target =nums.size()-k;
        int left=0;int right=nums.size()-1;
        while(true){
            int pivotIndex=partition(nums,left,right);
            if(pivotIndex==target){
                return nums[pivotIndex];
            }else if(pivotIndex<target){
                left=pivotIndex+1;
            }else{
                right=pivotIndex-1;
            }
        }
    }

    int partition(vector<int>&nums,int left,int right){
        int pivotIndex = left + rand() % (right - left + 1);//随机化pivot，这样能够避免快速排序的最坏情况的发生，最坏情况下时间复杂度高达o(n^2)
        swap(nums[pivotIndex], nums[right]);
        int pivot=nums[right];
        int i=left;
        for(int j=left;j<right;++j){
            if(nums[j]<pivot){
                swap(nums[i],nums[j]);
                i++;
            }
        }
        swap(nums[i],nums[right]);
        return i;
    }
};
```
### 计数排序+桶排序的思想
```c++
class Solution {
public:
    int findKthLargest(vector<int>& nums, int k) {
        const int OFFSET = 10000;
        const int SIZE = 20001;
        vector<int> cnt(SIZE, 0);
        for (int x : nums) {
            cnt[x + OFFSET]++;
        }
        int sum = 0;
        for (int i = SIZE - 1; i >= 0; --i) {
            sum += cnt[i];
            if (sum >= k) {
                return i - OFFSET;
            }
        }
        return -1;
    }
};
```

### 基数排序
```c++
class Solution {
public:
    void radixSort(vector<int>& nums){
        int offset = 10000;
        for(int& x : nums){
            x += offset;
        }
        int maxVal = *max_element(nums.begin(), nums.end());
        for(int exp = 1; maxVal / exp > 0; exp *= 10){
            vector<int> output(nums.size());
            vector<int> count(10,0);
            for(int x : nums){
                count[(x/exp)%10]++;
            }
            for(int i = 1; i < 10; i++){
                count[i] += count[i-1];
            }
            for(int i = nums.size()-1; i >= 0; i--){
                int digit = (nums[i]/exp)%10;
                output[count[digit]-1] = nums[i];
                count[digit]--;
            }
            nums = output;
        }
        for(int& x : nums){
            x -= offset;
        }
    }
    int findKthLargest(vector<int>& nums, int k){
        radixSort(nums);
        return nums[nums.size()-k];
    }
};
```
基数排序需要和计数排序一块用，大for循环中的三个小for循环就是在统计每个余数出现的次数，第一个for循环是统计每种余数的个数，第二个for循环是统计从小到大每个余数的第一个应该出现的位置，第三个for循环是将数组中的元素按余数放置到对应位置去

题干：给定单个链表的头 head ，使用 插入排序 对链表进行排序，并返回 排序后链表的头。  
插入排序 算法的步骤:  
1. 插入排序是迭代的，每次只移动一个元素，直到所有元素可以形成一个有序的输出列表。
2. 每次迭代中，插入排序只从输入数据中移除一个待排序的元素，找到它在序列中适当的位置，并将其插入。
3. 重复直到所有输入数据插入完为止。  

下面是插入排序算法的一个图形示例。部分排序的列表(黑色)最初只包含列表中的第一个元素。每次迭代时，从输入数据中删除一个元素(红色)，并就地插入已排序的列表中。  
```c++
/**
 * Definition for singly-linked list.
 * struct ListNode {
 *     int val;
 *     ListNode *next;
 *     ListNode() : val(0), next(nullptr) {}
 *     ListNode(int x) : val(x), next(nullptr) {}
 *     ListNode(int x, ListNode *next) : val(x), next(next) {}
 * };
 */
class Solution {
public:
    ListNode* insertionSortList(ListNode* head) {
        if (!head) return head;
        ListNode* dummy = new ListNode(0);
        ListNode* cur = head;
        while (cur) {
            ListNode* next = cur->next;
            ListNode* prev = dummy;
            while (prev->next && prev->next->val < cur->val) {
                prev = prev->next;
            }
            cur->next = prev->next;
            prev->next = cur;
            cur = next;
        }
        return dummy->next; 
    }
};
```

## 题目一 二分查找（没事就看看，不然最简单的二分法都忘记了）
题干：给定一个 n 个元素有序的（升序）整型数组 nums 和一个目标值 target  ，写一个函数搜索 nums 中的 target，如果 target 存在返回下标，否则返回 -1。  
你必须编写一个具有 O(log n) 时间复杂度的算法。
```c++
class Solution {
public:
    int search(vector<int>& nums, int target) {
        int left=0;
        int right=nums.size()-1;
        return binarySearch(nums,left,right,target);
    }
    
    int binarySearch(vector<int>&nums,int left,int right,int target){
        if(left>right){return -1;}
        if(nums[(left+right)/2]==target){
            return (left+right)/2;
        }else if(nums[(left+right)/2]>target){
            return binarySearch(nums,left,((left+right)/2-1),target);
        }else{
            return binarySearch(nums,((left+right)/2+1),right,target);
        }
    }
};
//非递归形式的代码
int binarySearch(vector<int>& nums, int target){
    int left = 0;
    int right = nums.size()-1;
    while(left <= right){
        int mid = left + (right-left)/2;
        if(nums[mid] == target)
            return mid;
        else if(nums[mid] < target)
            left = mid + 1;
        else
            right = mid - 1;
    }
    return -1;
}
```

## 题目二 山脉数组的峰顶索引
题干：给定一个长度为 n 的整数 山脉 数组 arr ，其中的值递增到一个峰值元素然后递减。返回峰值元素的下标。你必须设计并实现时间复杂度为 O(log(n)) 的解决方案。
### 错误代码
```c++
class Solution {
public:
    int peakIndexInMountainArray(vector<int>& arr) {
        int Max=0;
        for(int n:arr){
            Max=max(n,Max);
        }
        return binarySearch(arr,0,arr.size()-1,Max);
    }

    int binarySearch(vector<int>& arr,int left,int right,int target){
        if(left>right){return -1;}
        int mid=(left+right)/2;
        if(arr[mid]==target){
            return mid;
        }else if(arr[mid]>target){
            return binarySearch(arr,left,mid-1,target);
        }else{
            return binarySearch(arr,mid+1,right,target);
        }
    }
};
```
二分查找只能在有序数组中使用，这个错误的代码就是忽略了这个问题，所以最终必然不能通过用例
### 正确代码
其实这题的关键在于如何理解“山”这个概念，山峰：这个点左侧和右侧均比它低；上坡：这个点左侧比它低，右侧比它高，山峰在它的右侧；下坡：这个点左侧比它高，右侧比它低，山峰在它的左侧
```c++
class Solution {
public:
    int peakIndexInMountainArray(vector<int>& arr) {
        int left=0;
        int right=arr.size()-1;
        int mid=0;
        while(left<right){
            mid=(left+right)/2;
            if(arr[mid]>arr[mid+1]&&arr[mid]>arr[mid-1]){
                break;
            }else if(arr[mid]>arr[mid+1]&&arr[mid]<arr[mid-1]){
                right=mid;
            }else if(arr[mid]<arr[mid+1]&&arr[mid]>arr[mid-1]){
                left=mid;
            }
        }
        return mid;
    }
};
//上面那个代码其实还能优化，其实不用和左右都比，只用和左边比判断是在上升还是下降即可，优化后代码如下：
class Solution {
    public int peakIndexInMountainArray(int[] arr) {
        int left=0;
        int right=arr.length-1;
        int mid=0;
        while(left<right){
            mid=left+(right-left)/2;
            //左右都小于mid，说明mid是山峰。
            if(arr[mid-1]<arr[mid]&& arr[mid]>arr[mid+1]) break;
            //右边比左边高，说明山峰在右侧
            if(arr[mid+1]>arr[mid]) left=mid;
            //右边比左边低，山峰在左侧
            else if(arr[mid+1]<arr[mid]) right=mid;
        }
        return mid;
    }
}
```
为什么有些题是 right = mid，有些是 right = mid - 1，原因如下：  
核心其实只有一句话：mid 这个位置有没有可能是答案？
- 如果 可能是答案 → right = mid
- 如果 一定不是答案 → right = mid - 1

但是为什么left = mid + 1 几乎从来不会写成 left = mid，原因如下：
二分查找每次循环必须做到：搜索区间严格缩小，所以更新区间时必须保证：left 向右移动 或 right 向左移动  

为什么有些二分的参考代码中，mid的定义是**left+(right-left)/2**（写这种更好），这种定义和(left+right)/2有什么区别，又有什么好处  
两种写法在数学上是一样的，但在程序实现上有一个关键差别：防止整数溢出。两个极大的数相加后是有可能溢出的，但是先做差是一定不会溢出的。

题干：给定一个非负整数 c ，你要判断是否存在两个整数 a 和 b，使得 a^2 + b^2 = c 。
```c++
class Solution {
public:
    bool judgeSquareSum(int c) {
        long long left=0;
        long long right=sqrt(c);
        while(left<=right){
            long long num=left*left+right*right;
            if(num==c){
                return true;
            }else if(num>c){
                right--;
            }else{
                left++;
            }
        }
        return false;
    }
};
```

## 搜索旋转排序数组
题干：整数数组 nums 按升序排列，数组中的值 互不相同 。  
在传递给函数之前，nums 在预先未知的某个下标 k（0 <= k < nums.length）上进行了 向左旋转，使数组变为 [nums[k], nums[k+1], ..., nums[n-1], nums[0], nums[1], ..., nums[k-1]]（下标 从 0 开始 计数）。例如， [0,1,2,4,5,6,7] 下标 3 上向左旋转后可能变为 [4,5,6,7,0,1,2] 。  
给你 旋转后 的数组 nums 和一个整数 target ，如果 nums 中存在这个目标值 target ，则返回它的下标，否则返回 -1 。  
你必须设计一个时间复杂度为 O(log n) 的算法解决此问题。
### 两次二分方法
```c++
class Solution {
public:
    int search(vector<int>& nums, int target) {
        int left=0;
        int right=nums.size()-1;
        int mid=0;
        while(left<right){
            mid=left+(right-left)/2;
            if(nums[mid]>nums[right]){
                left=mid+1;
            }else{
                right=mid;
            }
        }
        int pivot=left;
        if(target>=nums[pivot]&&target<=nums[nums.size()-1]){
            left=pivot;
            right=nums.size()-1;
        }else{
            left=0;
            right=pivot-1;
        }
        while(left<=right){
            int mid=left+(right-left)/2;
            if(nums[mid] == target) {
                return mid;
            }else if(nums[mid]>target){
                right=mid-1;
            }else{
                left=mid+1;
            }
        }
        return -1;
    }
};
```
查找目标元素一般就是用二分法，在有序数组用二分法其实很简单，这个题目是两个有序数组的拼接，找到它们的分界点即可，但是其和山峰数组找山峰有所不同，所以这道题的关键就是如何使用二分法找到这个数组的分界点，一次遍历可以但是时间复杂度为o(n)，即使你后面在有序数组中找目标元素用二分法控制时间复杂度这也是超的，所以找分界点也得用二分法，二分法的精髓就是找一个点用于比较，进而判断下一次应该去哪里寻找我的目标，观察这个数组的结构可以发现，数组由于最开始是单增的，旋转后的后半部分的末尾是和前半部分的头恰好就是可以用于判断的条件，以后半部分末尾为例，即如果你这个值大于末尾的值，证明你落在了前半部分，因为前半部分的所有值都会大于末尾值，同时如果你小于末尾值了，那么你肯定落在了后半部分，那么根据这个条件我们可以控制二分点不断向数组的分界点去逼近，即前半部分和后半部分的交汇处去逼近
### 一次二方法
关键点：旋转数组虽然整体无序，但局部一定有一半有序；每次利用这半边有序区间判断 target 应该去哪边。如果target在有序部分的范围内那么就去有序部分继续二分查找就行了，如果不在这个范围内，那就去无序部分，在对无序部分划分为有序和无序部分，如何就又可以去判断是否在有序部分内了
```c++
class Solution {
public:
    int search(vector<int>& nums, int target) {
        int left = 0;
        int right = nums.size() - 1;
        while (left <= right) {
            int mid = left + (right - left) / 2;
            if (nums[mid] == target) {return mid;}
            if (nums[left] <= nums[mid]) {// 左半边有序
                if (nums[left] <= target && target < nums[mid]) {//在有序部分内，搜索范围缩小到left和mid-1
                    right = mid - 1;
                } else {//不在有序部分内，搜索范围缩小到mid+1和right
                    left = mid + 1;
                }
            }
            else {// 右半边有序
                if (nums[mid] < target && target <= nums[right]) {
                    left = mid + 1;
                } else {
                    right = mid - 1;
                }
            }
        }
        return -1;
    }
};
```

## 二分查找总结
1. 核心思想
    - 二分查找的本质：利用数据或答案的单调性，每次排除一半搜索空间
    - 基本流程

        确定搜索区间 [left, right]
        ↓
        计算中点 mid
        ↓
        比较 mid 与**目标**的关系（这个“目标”是要视情况而定的）
        ↓
        排除一半区间
        ↓
        重复直到找到**答案**(答案不一定全是mid，也有可能是left或者right)
2. 二分查找适用条件

    二分查找适用于存在单调性的问题
    - 有序数组
    - 单峰结构
    - 旋转数组
    - 答案具有单调性
3. 二分查找的区间模型

    - 闭区间[left,right]
        - 循环条件while(left <= right)
        - 更新方式left = mid + 1;right = mid - 1;
        - 适用于精确查找元素
    - 左闭右开[left,right)
        - 循环条件while(left < right)
        - 更新条件left = mid + 1;right = mid;
        - 适用于边界查找、二分答案
4. 常见二分查找模板
    - 普通二分（查找元素）
    - 左边界查找
        - 找第一个>=target的位置
        ```c++
        int left_bound(vector<int>& nums,int target){
            int left=0,right=nums.size()-1;
            while(left<=right){
                int mid=left+(right-left)/2;
                if(nums[mid] >= target)
                    right = mid - 1;
                else
                    left = mid + 1;
            }
            return left;
        }
        ```
    - 右边界查找
        - 找最后一个<=target的位置
        ```c++
        int right_bound(vector<int>& nums,int target){
            int left=0,right=nums.size()-1;
            while(left<=right){
                int mid=left+(right-left)/2;
                if(nums[mid] <= target)
                    left = mid + 1;
                else
                    right = mid - 1;
            }
            return right;
        }
        ```
    - 峰值查找
        - 适用于先升后降数组
        ```c++
        while(left < right){
            int mid = left + (right-left)/2;
            if(nums[mid] < nums[mid+1])
                left = mid + 1;
            else
                right = mid;
        }
        return left;
        ```
    - 二分答案
        - 结构：判断mid是否满足条件
        ```c++
        while(left < right){
            int mid = left + (right-left)/2;
            if(check(mid))
                right = mid;
            else
                left = mid + 1;
        }
        return left;
        ```
5. 二分法的常见错误
    - mid溢出，建议写mid = left + (right-left)/2
    - 死循环（区间没有有效缩小），错误写法left = mid;right = mid;
    - 循环条件错误导致漏掉答案，例：循环用while(left < right)，但right = mid - 1
    - 未处理找不到的情况，最终必须有一个return -1;

## 补：什么是二分答案
核心思想：二分答案 = 在一个具有单调性的答案空间中，利用二分查找寻找满足条件的最优解。  
二分答案是一类特殊的二分查找问题，其核心思想不是在一个数组中查找某个元素，而是**在一个可能的答案范围中寻找满足条件的最优解**。通常题目会要求我们求一个“最小值”或“最大值”，并且存在一个**判断函数**，可以**判断某个候选答案是否满足条件**。当这个**判断结果随着答案的变化呈现出单调性时**，就可以使用二分查找来快速定位答案。
这类问题通常具有如下结构：随着某个参数 x 的增大或减小，问题的可行性会发生变化。例如可能出现这样的情况：当 x 较小时问题无法满足条件，而当 x 足够大时就可以满足条件。这样一来，判断函数的结果就会呈现出类似下面的结构：  
false false false false true true true true  
                    ↑  
                最小可行值  
也就是说，在某个位置之前全部不满足条件，而在这个位置之后全部满足条件。二分答案的目标就是找到这个第一个满足条件的位置。
- 题目示例

LeetCode 875《爱吃香蕉的珂珂》。题目给定若干堆香蕉，例如piles = [3,6,7,11]，同时给定一个时间限制：h = 8。珂珂每小时可以选择一堆香蕉吃，并且每小时最多吃k根。如果这一堆不足 k，则全部吃掉。题目要求求出一个最小的整数 k，使得珂珂能够在 h小时内吃完所有香蕉。如果我们尝试不同的速度 k，可以得到如下结果：  
| k | 是否能在 8 小时内吃完 |
| - | :------------: |
| 1 | 否            |
| 2 | 否            |
| 3 | 否            |
| 4 | 是            |
| 5 | 是            |
| 6 | 是            |  

可以看到，当 $k<4$ 时都无法完成，而从 $k=4$ 开始就都可以完成。因此答案就是最小的可行速度4。这正好符合“前面全部不满足，后面全部满足”的单调结构，因此可以使用二分查找。  
要使用二分答案，首先需要确定答案的搜索范围。在这个问题中，速度 k 的最小值是 1，而最大值不会超过最大的一堆香蕉数量，因此范围是：$1≤k≤max(piles)$ 接下来需要定义一个判断函数 check(k)，用于判断当吃香蕉的速度为 k时，是否能够在 h小时内吃完所有香蕉。计算方法是统计每一堆香蕉需要的时间，然后求和。如果总时间不超过 h，说明当前速度是可行的。  
在这个过程中，二分查找不断缩小答案区间。如果某个速度 mid 可以在规定时间内完成任务，则说明真正的最小速度可能更小，因此需要向左继续搜索；如果当前速度无法完成任务，则说明速度太慢，需要向右搜索更大的速度。最终，当区间收缩到一个点时，这个点就是最小可行答案。  
总结来说，二分答案的关键在于将问题转化为“寻找第一个满足条件的位置”。与普通二分查找不同，普通二分查找是在数组中寻找某个具体元素，而二分答案是在一个数值范围内寻找最优解。其核心步骤通常包括三个部分：首先确定答案的搜索区间，其次设计一个判断函数 check(x) 用来判断某个答案是否可行，最后通过二分查找找到满足条件的最小或最大值。  
很多算法题都可以转化为这种形式，例如“最小速度”“最小时间”“最大距离”“最小代价”等问题。只要题目中存在明确的单调关系，并且可以写出一个判断函数，就很可能可以使用二分答案来解决。

## 题目一
题干：如果长度为 n 的数组 nums 满足下述条件，则认为该数组是一个漂亮数组：
- nums 是由范围 [1, n] 的整数组成的一个排列。
- 对于每个 0 <= i < j < n ，均不存在下标 k（i < k < j）使得 2 * nums[k] == nums[i] + nums[j] 。
给你整数 n ，返回长度为 n 的任一 漂亮数组 。本题保证对于给定的 n 至少存在一个有效答案。
```c++
class Solution {
public:
    vector<int> beautifulArray(int n) {
        vector<int>result{1};
        while(result.size()<n){
            vector<int>temp;
            for(int i=0;i<result.size();++i){
                if(2*result[i]-1<=n){
                    temp.push_back(2*result[i]-1);
                }
            }
            for(int i=0;i<result.size();++i){
                if(2*result[i]<=n){
                    temp.push_back(2*result[i]);
                }
            }
            result=temp;
        }
        return result;
    }
};
```
本题需要一定的数学敏感度，从题目中的那个表达式出发，首先等式左边一定是一个偶数，那么什么情况下这个表达式一定不成立就很明了了，即右边是一个偶数加一个奇数，所有可以将奇数全放在左边，偶数全放在右边，这时候有一个新问题来了，奇数偶数分开放，如果全在奇数侧取或者全在偶数侧取，不是还有可能在形式上满足上述表达式吗，但是这时候有另外一个约束，即对漂亮数组全乘和全加是不会改变其漂亮数组的性质的，而我们放到左侧去的奇数组是在漂亮数组的基础上构建出来的，右侧的偶数组也是如此，所以你即使全取在了奇数侧或偶数侧也仍然是漂亮数组

## 题目二 从中序与后序遍历序列构造二叉树
题干：给定两个整数数组 inorder 和 postorder ，其中 inorder 是二叉树的中序遍历， postorder 是同一棵树的后序遍历，请你构造并返回这颗二叉树 。
### 基本代码（能够正确实现功能，但是当中序和后续数组很大的时候会超时）
```c++
/**
 * Definition for a binary tree node.
 * struct TreeNode {
 *     int val;
 *     TreeNode *left;
 *     TreeNode *right;
 *     TreeNode() : val(0), left(nullptr), right(nullptr) {}
 *     TreeNode(int x) : val(x), left(nullptr), right(nullptr) {}
 *     TreeNode(int x, TreeNode *left, TreeNode *right) : val(x), left(left), right(right) {}
 * };
 */
class Solution {
public:
    TreeNode* buildTree(vector<int>& inorder, vector<int>& postorder) {
        if(inorder.size()==0||postorder.size()==0){return nullptr;}
        return split(inorder,postorder);
    }

    int findIndex(vector<int>& nums,int target){
        for(int i=0;i<nums.size();++i){
            if(nums[i]==target){return i;}
        }
        return -1;
    }

    TreeNode* split(vector<int>&inorder,vector<int>&postorder){
        if(inorder.size()==0){return nullptr;}
        int rootVal=postorder[postorder.size()-1];
        TreeNode* rootNode=new TreeNode(rootVal);
        if(postorder.size()==1){return rootNode;}

        vector<int>leftInorder;
        vector<int>rightInorder;
        int splitIndex=findIndex(inorder,rootVal);
        for(int i=0;i<inorder.size();++i){
            if(i<splitIndex){leftInorder.push_back(inorder[i]);}
            else if(i>splitIndex){rightInorder.push_back(inorder[i]);}
        }

        vector<int>leftPostorder;
        vector<int>rightPostorder;
        for(int j=0;j<postorder.size()-1;++j){//时间复杂度o(n^2)
            if(findIndex(leftInorder,postorder[j])>=0){
                leftPostorder.push_back(postorder[j]);
            }
            else if(findIndex(rightInorder,postorder[j])>=0){
                rightPostorder.push_back(postorder[j]);
            }
        }

        rootNode->left=split(leftInorder,leftPostorder);
        rootNode->right=split(rightInorder,rightPostorder);
        return rootNode;
    }
};
```
### 优化后代码
```c++
/**
 * Definition for a binary tree node.
 * struct TreeNode {
 *     int val;
 *     TreeNode *left;
 *     TreeNode *right;
 *     TreeNode() : val(0), left(nullptr), right(nullptr) {}
 *     TreeNode(int x) : val(x), left(nullptr), right(nullptr) {}
 *     TreeNode(int x, TreeNode *left, TreeNode *right) : val(x), left(left), right(right) {}
 * };
 */
class Solution {
public:
    unordered_map<int,int>mp;

    TreeNode* buildTree(vector<int>& inorder, vector<int>& postorder) {
        int n=inorder.size();
        for(int i=0;i<n;++i){
            mp[inorder[i]]=i;
        }
        return split(inorder,postorder,0,n-1,0,n-1);
    }

    TreeNode* split(vector<int>&inorder,vector<int>&postorder,int inL,int inR,int postL,int postR){
        if (inL > inR || postL > postR) return nullptr;
        int rootVal=postorder[postR];
        TreeNode* rootNode=new TreeNode(rootVal);

        int k=mp[rootVal];
        int leftSize=k-inL;

        rootNode->left=split(inorder,postorder,inL,k-1,postL,postL+leftSize-1);
        rootNode->right=split(inorder,postorder,k+1,inR,postL+leftSize,postR-1);
        return rootNode;
    }
};
```
优化点主要在查找索引位置这块，利用哈希表将其时间复杂度压为了o(1),避免了之前那个代码里面时间复杂度为o(n^2)的地方，而且不用额外存储左/右子树的中序/后序数组了

## 题目三 翻转对
题干：给定一个数组 nums ，如果 i < j 且 nums[i] > 2*nums[j] 我们就将 (i, j) 称作一个重要翻转对。  
你需要返回给定数组中的重要翻转对的数量。
### 错误的排序思想
我希望将所有元素的一半放入数组中，然后排序，看有多少个“一半值”大于“原值”，但是这样忽视了一个条件就是i < j，最终结果也必然错了
```c++
class Solution {
public:
    int reversePairs(vector<int>& nums) {
        vector<pair<double,bool>> store;
        for(int i=0;i<nums.size();++i){
            store.push_back({nums[i],0});
            store.push_back({nums[i]/2.0,1});
        }
        sort(store.begin(),store.end());
        int n = store.size();
        int ans = 0;
        for (int i = 0; i < n; ) {
            if (store[i].second == 0) {
                i++;
                continue;
            }
            int j = i;
            while (j < n && store[j].second == 1) {
                j++;
            }
            int left = i - 1;
            int right = j;
            if (left >= 0 && right < n && store[left].second == 0 && store[right].second == 0) {
                ans += (j - i);
            }
            i = j;
        }
        return ans;
    }
};
```
### 归并排序思想下的算法（本质是分治思想，归并就是一种分治）
```c++
class Solution {
public:
    int reversePairs(vector<int>& nums) {
        return mergeSort(nums,0,nums.size()-1);
    }

    int mergeSort(vector<int>&nums,int left,int right){
        if(left>=right)return 0;
        int mid=left+(right-left)/2;
        int ans=0;
        ans+=mergeSort(nums,left,mid);
        ans+=mergeSort(nums,mid+1,right);
        ans+=merge(nums,left,mid,right);
        return ans;
    }

    int merge(vector<int>&nums,int left,int mid,int right){
        int ans=0;
        int i=left;int j = mid + 1;
        for (i = left; i <= mid; ++i) {
            while (j <= right && (long long)nums[i] > 2LL * nums[j]) {
                ++j;
            }
            ans += j - (mid + 1);
        }

        vector<int>temp;
        i=left;j=mid+1;
        while(i<=mid&&j<=right){
            if(nums[i]<=nums[j]){
                temp.push_back(nums[i]);++i;
            }else{
                temp.push_back(nums[j]);++j;
            }
        }
        while(i<=mid){
            temp.push_back(nums[i]);++i;
        }
        while(j<=right){
            temp.push_back(nums[j]);++j;
        }
        for(int k=0;k<temp.size();++k){
            nums[left+k]=temp[k];
        }
        return ans;
    }
};
```
这道题要求统计满足 i < j 且 nums[i] > 2 * nums[j] 的数对数量。本质上这是一个“逆序对”的变形问题，不仅需要满足大小关系，还需要满足下标顺序约束。与普通逆序对不同的是，这里的比较条件变成了“左边元素大于右边元素的两倍”，这使得问题难度有所提升。  
最直接的做法是使用双重循环枚举所有 (i, j) 对，并判断是否满足条件。这种方法时间复杂度为 O(n^2)，在数据规模较大时会超时。因此需要寻找更高效的算法来优化这一过程。  
这道题可以借助归并排序的思想进行优化。归并排序的核心是“分治”，即将数组不断拆分成更小的子数组，分别排序后再合并。在这个过程中，可以顺便完成“统计数对”的任务。具体来说，每次递归会将数组划分为左右两个区间 [left, mid] 和 [mid+1, right]，并分别保证这两个区间在递归结束后是有序的。  
归并排序的关键在于，递归完成后，我们不仅可以得到左区间内部和右区间内部的答案，还可以利用“左右区间已经有序”这一性质，高效统计跨区间的数对数量，也就是满足 i 在左区间，j 在右区间 的情况。这一步是整个算法的核心。  
当左右区间已经有序时，对于左区间中的某个元素 nums[i]，右区间中满足 nums[i] > 2 * nums[j] 的元素一定构成一个连续的前缀。这是因为右区间是递增的，一旦某个位置不满足条件，后面的元素也一定不满足。因此，我们可以用一个指针 j 从右区间的起点开始移动，对于每个 i，不断向右推进 j，直到条件不再满足。此时，满足条件的元素个数就是 j - (mid + 1)。  
由于左区间本身也是有序的，随着 i 的增大，满足条件的右区间范围只可能保持不变或向右扩展，因此指针 j 不需要回退，可以整体保持线性移动。这一点非常关键，它保证了跨区间统计的时间复杂度是 O(n)，从而使整个算法的复杂度维持在 O(n log n)。  
在完成跨区间统计之后，还需要进行标准的归并操作，将左右两个有序区间合并成一个更大的有序区间。这样做的目的是保证当前区间整体有序，以便在上一层递归中继续利用这一性质进行统计。归并的过程与普通归并排序完全一致，使用双指针依次比较两个区间的元素，将较小值放入临时数组中，最后再写回原数组对应位置。  
整个递归过程的结构可以总结为三部分：首先递归计算左区间的答案，其次递归计算右区间的答案，最后计算跨区间的答案，并将三者相加得到当前区间的总答案。在此过程中，归并排序不仅完成了排序任务，还通过“有序性”帮助我们高效统计了满足条件的数对。 
 
这个题目的思考路径应该是这样的：
1. 这是一个 pair 问题：i < j
2. 暴力是 O(n^2) → 太慢
3. 转换为：对每个 i，找右边有多少 j 满足条件
4. 如果右边有序，就能快速统计
5. 怎么让右边有序？

关键在于能够思考到**右边有序的时候是能够快速统计的**，即如果右边是递增的，那么我只用找到第一次不满足题目所给不等式的位置即可，因为继续向右走只会更大，肯定更不可能满足条件了，这个思想在归并统计的时候也用到了，左侧i和右侧j有多少个满足条件的数对，同时当i改变后j不需要置为0，因为i变大，要满足不等式j只能更大，所有j直接继续向后就行了，这个操作也将查找操作的复杂度给降到了o(n),而不是o(n^2)。那么想到上述内容后我们要去思考的就是如何实现右侧的有序，然后顺着这个思路继续往下想就行

### 树状数组
1. 什么是树状数组(用多个“不同大小的区间”拼出前缀)(用二进制把区间拆成若干个 2^k 的块)

树状数组是一种用来高效维护“前缀信息”的数据结构，最经典的用途是：单点更新、前缀和查询。比如你有一个数组 a，树状数组可以支持：把某个位置加上一个值、快速求 a[1] + a[2] + ... + a[x]。这两个操作的时间复杂度都是：O(log n)，而普通数组如果每次都从头累加，查询前缀和要 O(n)。

2. 适用的问题

假设你一边遍历数组，一边不断插入数字，然后经常问：当前已经出现的数中，有多少个 <= x？这类“动态计数 + 前缀统计”问题，就是树状数组很擅长的场景。

3. 树状数组的本质

一个专门用来维护前缀频率和前缀和的结构。例如：  
我们把某些值离散化后映射到下标：  
值:    1   3   5   8  
下标:  1   2   3   4  
然后树状数组里存的是：值 1 出现了几次、值 3 出现了几次、值 5 出现了几次、值 8 出现了几次。这样你就能快速问：小于等于 5 的一共有多少个、小于 8 的一共有多少个，本质就是“频率表 + 前缀和”。

4. 树状数组的两个核心操作

    - 单点更新add

    含义：给位置 x 增加 delta、然后跳到它影响的更大区间
    ```c++
    void add(int x, int delta) {
        while (x <= n) {
            tree[x] += delta;
            x += x & -x;
        }
    }
    ```
    - 前缀和查询query

    含义：求前缀 [1, x] 的总和
    ```c++
    int query(int x) {
        int res = 0;
        while (x > 0) {
            res += tree[x];
            x -= x & -x;
        }
        return res;
    }
    ```

    - x & -x是什么操作

    表示取出x的二进制表示中最低位的1对应的值，例如：6 = 110，6 & -6 = 2 |||| 8 = 1000，8 & -8 = 8。  
    树状数组中，每个节点负责一段区间，而这个区间长度就是x & -x  
    例如，当n=8时  
    | 下标 x | x & -x | 负责区间  |
    | :----: | :------: | :-----： |
    | 1    | 1      | [1]   |
    | 2    | 2      | [1,2] |
    | 3    | 1      | [3]   |
    | 4    | 4      | [1,4] |
    | 5    | 1      | [5]   |
    | 6    | 2      | [5,6] |
    | 7    | 1      | [7]   |
    | 8    | 8      | [1,8] |
    tree[x] 管的是：[x - (x & -x) + 1, x]  

    - 为什么query用x -= x & -x

        例如query(6)
        - 第一步：
        ```c++
        res += tree[6]   // 管 [5,6]
        x -= 2 → x = 4
        ```
        - 第二步：
        ```c++
        res += tree[4]   // 管 [1,4]
        x -= 4 → x = 0
        ```
        - 覆盖区间：[5,6] + [1,4] = [1,6]，刚好是前缀

    - 为什么add用x += x & -x

        给a[x]加1，那么所有包含x的区间都需要进行更新  
        例如tree[3]+1
        - 第一步：
        ```c++
        x += 1 → 4
        tree[4] += 1   // 管 [1,4]
        ```
        - 第二步：
        ```c++
        x += 4 → 8
        tree[8] += 1   // 管 [1,8]
        ```

5. 到底怎么理解树状数组

树状数组并**不是“把原数组直接变成前缀和数组”**。它存的**不是“每个位置从 1 到 i 的完整前缀和”**，而是一些**局部区间和**。更准确地说，tree[x] 存储的是一个固定区间的和，这个区间是：[x - lowbit(x) + 1, x]，其中 lowbit(x) = x & -x。比如下标从 1 开始时：tree[1] 存 [1,1]、tree[2] 存 [1,2]、tree[3] 存 [3,3]、tree[4] 存 [1,4]、tree[5] 存 [5,5]、tree[6] 存 [5,6]、tree[7] 存 [7,7]、tree[8] 存 [1,8]。所以树状数组存的不是“所有前缀和”，而是若干个长度为 2 的幂的区间和。它的厉害之处在于：这些区间可以像积木一样拼出任意前缀 [1, x]。  

你担心“如果它存的是前缀和，原数组信息不是丢了吗”，这个担心是合理的，但这里要分清楚：树状数组通常不是拿来“还原整个原数组结构”的，它的目标是高效支持两种操作：单点更新、前缀查询。只要这两个操作能正确完成，它就够用了。并且原数组信息其实没有完全丢，因为**单个元素也可以通过前缀差求出来**：**a[x] = prefix(x) - prefix(x-1)**，也就是说，只要树状数组能查前缀和，就仍然能间接得到原数组某个位置的值。  

树状数组的长度与原数组的长度是相同的，因为其要保证每个位置i都能作为“区间的右端点”

为什么最低位1的值刚好表示区间长度？？lowbit(x) = x & -x = “x 能整除的最大 2 的幂”。换个角度理解：它其实是“2 的幂分解”，任何整数都可以写成x = (某个数) × 2^k，其中 2^k 就是它能整除的最大 2 的幂。其实就是“二进制分解(把一个数拆成若干个2的幂的和)”，例如13=1101，即13=8+4+1= 2^3 + 2^2 + 2^0。那么这和区间长度有什么关系呢？现在我们要解决一个问题：如何把区间 [1, x] 拆成若干个“规则的小区间”？要求：不重叠、能拼回原区间、每块结构统一（方便计算），那么这个二进制分解出来的结果就刚好完美符合，我们取分解最后一个数作为区间长度就刚刚好，因为这样就能保证不重不漏了，那么这个数刚好对应了二进制的最低位1的值

为什么x & -x操作刚好能够取出最低位1呢？首先计算机存的是补码，正数补码与原码没区别，但是负数补码是原码按位取反再加1，你加进去的这个1只能刚好作用到原码最低位1的那个位置，因为这个位置取反后是0，加1不会再进位了，但是最低位1后面那些原本为0的地方，取反后变成1，加1会产生进位操作

普通前缀和是“存答案”，树状数组是“存构成答案的零件”

### 树状数组求解
```c++
class Solution {
public:
    struct BIT {
        int n;
        vector<int> tree;
        BIT(int n) : n(n), tree(n + 1, 0) {}
        void add(int x, int delta) {
            while (x <= n) {
                tree[x] += delta;
                x += x & -x;
            }
        }
        int query(int x) {
            int res = 0;
            while (x > 0) {
                res += tree[x];
                x -= x & -x;
            }
            return res;
        }
    };
    int reversePairs(vector<int>& nums) {
        vector<long long> alls;
        for (int x : nums) {
            alls.push_back((long long)x);
            alls.push_back(2LL * x);
        }
        sort(alls.begin(), alls.end());
        alls.erase(unique(alls.begin(), alls.end()), alls.end());
        auto getRank = [&](long long x) {
            return (int)(lower_bound(alls.begin(), alls.end(), x) - alls.begin()) + 1;
        };
        BIT bit(alls.size());
        int ans = 0;
        for (int j = 0; j < nums.size(); ++j) {
            long long x = nums[j];
            int idx2 = getRank(2LL * x);

            int notGreater = bit.query(idx2);   // <= 2*x 的个数
            int total = j;                      // 左边已经插入了 j 个数
            ans += total - notGreater;          // > 2*x 的个数

            int idx = getRank(x);
            bit.add(idx, 1);
        }
        return ans;
    }
};
```
这里的思路就是利用树状数组能够动态求取前缀和的特性来实现对不等式的统计，树状数组是逐步扩大的，其中存的是元素出现的次数，这个元素是根据j的递增逐步存进去的，也就是i,后面j进来后在树状数组中query两倍的j的索引，就能够统计出小于两倍的j是数的个数，然后再用当前树状数组的大小做差就能得到大于两倍j的个数（因为树状数组更擅长往前求和，而不是向后求和），然后将j的索引存入数组供下一次查询使用  

在树状数组解法中，真正插入树状数组的只有原数组中的一倍值 nums[i]，因为树状数组维护的是“已经遍历过的元素的出现次数”。二倍值 2 * nums[j] 并不会被插入，它的作用只是作为查询时的上界。之所以离散化时要同时加入一倍值和二倍值，是为了让“插入的值”和“查询的边界”都映射到同一套离散坐标系中。这样在处理当前元素 nums[j] 时，虽然 2 * nums[j] 本身没有存入树状数组，但仍然可以通过 query(rank(2 * nums[j])) 统计出所有不大于该边界的已插入元素个数。因此，一倍值负责存储，二倍值负责定位查询边界。

## 题目一 环形链表
题干：给你一个链表的头节点 head ，判断链表中是否有环。  
如果链表中有某个节点，可以通过连续跟踪 next 指针再次到达，则链表中存在环。 为了表示给定链表中的环，评测系统内部使用整数 pos 来表示链表尾连接到链表中的位置（索引从 0 开始）。注意：pos 不作为参数进行传递 。仅仅是为了标识链表的实际情况。  
如果链表中存在环，则返回true。 否则，返回false。
### 双指针（快慢指针）
如果有环的话指针是能够走回来的，那么原本在前面的指针就有可能到后面去了，那么如果这个指针移动的更快，它就有可能追上原本在后面的指针了，就和在跑道上跑步一个逻辑
```c++
/**
 * Definition for singly-linked list.
 * struct ListNode {
 *     int val;
 *     ListNode *next;
 *     ListNode(int x) : val(x), next(NULL) {}
 * };
 */
class Solution {
public:
    bool hasCycle(ListNode *head) {
        if(head==nullptr||head->next==nullptr){return false;}
        ListNode*slow=head;
        ListNode*fast=head;
        while(fast!=nullptr&&fast->next!=nullptr){
            slow=slow->next;
            fast=fast->next->next;
            if(fast==slow){
                return true;
            }
        }
        return false;
    }
};
```

将链表节点存在哈希表中，利用哈希表的唯一性就能解决问题

## 题目二 最接近的三数之和
题干：给你一个长度为 n 的整数数组 nums 和 一个目标值 target。请你从 nums 中选出三个在 不同下标位置 的整数，使它们的和与 target 最接近。返回这三个数的和。假定每组输入只存在恰好一个解。  
```c++
class Solution {
public:
    int threeSumClosest(vector<int>& nums, int target) {
        sort(nums.begin(),nums.end());
        int bestSum=nums[0]+nums[1]+nums[2];
        for(int i=0;i<nums.size()-2;++i){
            int j=i+1;
            int k=nums.size()-1;
            while(j<k){
                int sum=nums[i]+nums[j]+nums[k];
                if(abs(sum-target)<abs(bestSum-target)){
                    bestSum=sum;
                }
                if(sum>target){
                    k--;
                }else{
                    j++;
                }
            }
        }
        return bestSum;
    }
};
```

## 题目三 神奇字符串
题干：神奇字符串 s 仅由 '1' 和 '2' 组成，并需要遵守下面的规则：将连续相同字符组 '1' 和 '2' 长度的序列连接起来会生成字符串 s 本身。  
s 的前几个元素是 s = "1221121221221121122……" 。如果将 s 中连续的若干 1 和 2 进行分组，可以得到 "1 22 11 2 1 22 1 22 11 2 11 22 ......" 。每组中 1 或者 2 的出现次数分别是 "1 2 2 1 1 2 1 2 2 1 2 2 ......" 。上面的出现次数正是 s 自身。  
给你一个整数 n ，返回在神奇字符串 s 的前 n 个数字中 1 的数目。
```c++
class Solution {
public:
    int magicalString(int n) {
        vector<int> nums={1,2,2};
        int head=2;
        while(nums.size()<n){
            int nextNum=3-nums.back();
            int count=nums[head];
            for(int i=1;i<=count;++i){
                nums.push_back(nextNum);
            }
            head++;
        }
        int ans=0;
        for(int i=0;i<n;++i){
            if(nums[i]==1){
                ans++;
            }
        }
        return ans;
    }
};
```
这个题之所以在双指针里面是因为官方题解是用i记录需要插入的个数，用j记录需要插入的位置，其实和这个解法也大差不差，但是这题的关键是初始位置要设在2，而不是0

# 双指针总结
## 双指针常用“类型”
- 第一种是对撞型：两边向中间逼近，解决有序配对、回文、最优逼近。
- 第二种是追及型：一个快一个慢，解决链表判环、找中点、固定间隔。
- 第三种是窗口型：左右边界维护连续区间，解决子数组、子串最值问题。
- 第四种是读写型：一个读，一个写，解决原地删除、去重、压缩、搬移。
- 第五种是归并型：两个序列同步推进，解决合并、交集、比较。

## 什么是双指针

双指针并不是一种单独的题型，而是一大类“**用两个位置变量协同推进**”的思维方法。它的核心不是“两个指针”这四个字，而是：通过两个位置之间的关系，避免重复枚举，从而把暴力解法优化掉。很多时候，暴力是两层循环，而双指针的本质就是：想办法**让第二层循环不要每次都从头开始**。  
**纠正**：下面这个其实不是双指针是暴力枚举，j是在i的基础上改变的
```c++
for (int i = 0; i < n; ++i) {
    for (int j = i + 1; j < n; ++j) {
        ...
    }
}
```
下面这个是双指针,j不是在i的基础上改变的，但是会因为i或者与i有关的东西而改变
```c++
int i=left;int j = mid + 1;
for (i = left; i <= mid; ++i) {
    while (j <= right && (long long)nums[i] > 2LL * nums[j]) {
        ++j;
    }
    ans += j - (mid + 1);
}
```

## 双指针的适用特征
- 数据结构是线性的
- 位置之间有顺序关系
- 条件随着指针移动具有单调性
- 目标是找区间、配对、去重、压缩、相遇等

## 左右夹逼型双指针
一个指针在左边，一个指针在右边，两个指针从两端向中间靠拢，因此也叫“对撞指针”。左右两端的信息能够帮助你决定下一步该移动谁。  
常见用途：
- 在有序数组中找两个数使其和等于target（如果 nums[left] + nums[right] < target，说明和太小，要增大，只能左指针右移，如果 nums[left] + nums[right] > target，说明和太大，要减小，只能右指针左移，相等则找到答案）

- 判断回文串（比较“对称位置”）

- 容器盛水问题（如果当前较短板在左边，那么移动右边没有意义，因为宽度减小了，而短板没变，面积不可能变更优；所以应该移动短板那一侧，尝试找到更高的板。）

- 三数之和中的内部两指针
- 删除两端不满足条件的元素
- 从两侧逼近最优答案

## 快慢指针

常见用途：
- 判断链表有无环（快指针一次走两步，慢指针一次走一步：无环：快指针先到空，有环：快指针一定追上慢指针，本质上是“环形跑道追及问题”）

- 找链表中点（fast 速度是 slow 的两倍。当 fast 走完整个链表时，slow 正好走了一半。）

- 找倒数第 k 个节点（先让 fast 走 k 步，然后 slow 和 fast 一起走，当 fast 到末尾时，slow 就在倒数第 k 个节点）

- 判断环入口
- 数组中寻找重复数（抽象成环）
- 原地去重、原地覆盖时的“读写不同速”

## 滑动窗口型双指针

常见用途：
- 最长/最短子数组、子串
- 包含某些字符的最小覆盖子串

- 无重复字符的最长子串（右指针不断加入字符，若出现重复，就移动左指针直到窗口重新合法，每一步更新最长长度，关键不在于“双指针”本身，而在于：窗口里维护什么信息？通常会配合：哈希表、计数数组、set / map一起使用。）

- 和至少为 target 的最短子数组（右指针不断扩张，让区间和变大，一旦满足条件，就尝试移动左指针缩小窗口，在缩小过程中更新最优答案，因为数组元素都是正数，所以窗口右扩时和只会增大，左缩时和只会减小，具有单调性。）

- 固定长度窗口统计
- 满足某种频率约束的区间问题

本质：维护一个连续区间，并动态调整它，使其始终满足某种条件。和普通暴力相比，区别在于：暴力：枚举所有区间，滑动窗口：窗口左右端都只向前走，不回头，所以复杂度通常是 O(n)。

## 同向双指针

这类双指针都从左往右走，但职责不同。**一个负责“扫描”，一个负责“写入”或“维护有效区间”**。
常见用途：
- 原地删除元素
```c++
int removeElement(vector<int>& nums, int val) {
    int slow = 0;
    for (int fast = 0; fast < nums.size(); ++fast) {
        if (nums[fast] != val) {
            nums[slow] = nums[fast];
            slow++;
        }
    }
    return slow;//当前已保留元素的数量
}
```

- 数组去重（**fast：不断扫描新元素，slow：指向“下一个该写入的位置”**，最后 slow 就是新数组长度。）
```c++
int removeDuplicates(vector<int>& nums) {
    if (nums.empty()) return 0;
    int slow = 1;
    for (int fast = 1; fast < nums.size(); ++fast) {
        if (nums[fast] != nums[fast - 1]) {//表示出现新元素了
            nums[slow] = nums[fast];
            slow++;
        }
    }
    return slow;//slow是数组长度
}
```

- 合并有序结果时的推进
- 原地压缩数组

- 把满足条件的元素移动到前面
```c++
void moveZeroes(vector<int>& nums) {
    int slow = 0;
    for (int fast = 0; fast < nums.size(); ++fast) {
        if (nums[fast] != 0) {
            swap(nums[slow], nums[fast]);
            slow++;//当前“非零元素区间”的右边界
        }
    }
}
```

## 前后间隔型双指针

两个指针之间维持某种固定关系。比如：fast 比 slow 永远领先 k 步，这种思路经常用来把“倒数”转换成“同步前进”。  
常见用途：

## 归并型双指针

常见用途：

## 题目一 存在重复元素II
题干：给你一个整数数组 nums 和一个整数 k ，判断数组中是否存在两个 不同的索引 i 和 j ，满足 nums[i] == nums[j] 且 abs(i - j) <= k 。如果存在，返回 true ；否则，返回 false 。
### 优化的双循环（我以为是滑动窗口但是并不是，悲）
```c++
class Solution {
public:
    bool containsNearbyDuplicate(vector<int>& nums, int k) {
        for(int i=0;i<nums.size();++i){
            int n=nums.size()-1;
            if(k+i<n){n=k+i;}
            for(int j=i+1;j<=n;++j){
                if(nums[i]==nums[j]){
                    return true;
                }
            }
        }
        return false;
    }
};
```

### 哈希表
最简单的方法，存元素值和元素索引
```c++
class Solution {
public:
    bool containsNearbyDuplicate(vector<int>& nums, int k) {
        unordered_map<int,int>mp;
        for(int i=0;i<nums.size();++i){
            if(mp.count(nums[i])&&i-mp[nums[i]]<=k){
                return true;
            }else{
                mp[nums[i]]=i;
            }
        }
        return false;
    }
};
```

### 滑动窗口
什么是滑动窗口？  
窗口是“连续移动”的，且会复用之前的结果，也就是说：不会为每个 i 重新计算，而是“在上一个状态基础上调整”  

我的写法：每次重新计算一个区间（重复工作）  
滑动窗口：在前一个区间基础上“微调（整体右移，只有首尾元素需要微调）”（复用结果）  

滑动窗口是典型的空间换时间，其实就是相当于将我之前那个找元素的第二层循环用查找高效的数据结构给替换了  
本质上是将原本对长度不超过 k 的区间进行线性重复扫描，改成用哈希集合维护该区间内元素的存在性信息。**窗口负责动态维护合法范围**，哈希集合负责**将“是否存在相同元素”的查找从 O(k) 降到平均 O(1)**，因此整体上是典型的用空间换时间。
```c++
class Solution {
public:
    bool containsNearbyDuplicate(vector<int>& nums, int k) {
        unordered_set<int> window;
        for (int i = 0; i < nums.size(); ++i) {
            if (window.count(nums[i])) {
                return true;
            }
            window.insert(nums[i]);
            if (window.size() > k) {
                window.erase(nums[i - k]);
            }
        }
        return false;
    }
};
```

## 题目二 包含所有三种字符的子字符串数目
题干：给你一个字符串 s ，它只包含三种字符 a, b 和 c 。请你返回 a，b 和 c 都 至少 出现过一次的子字符串数目。 
### 双指针优化（但是还是有多余的暴力循环导致最终超时）
```c++
class Solution {
public:
    int numberOfSubstrings(string s) {
        int n = s.size();
        if (n < 3) return 0;
        int i = 0, j = 2, ans = 0;
        while (i <= n - 3) {
            unordered_set<char> Set;
            for (int k = i; k <= j; ++k) {
                Set.insert(s[k]);
            }//滑动窗口的核心就是为了避免这种情况的出现，每次不用从头开始创建一个窗口的状态，而是在原有状态上修改才叫滑动窗口
            if (Set.count('a') && Set.count('b') && Set.count('c')) {
                ans += n - j;   // 不是 n-j+1
                i++;
                if (j < i + 2) j = i + 2;
            } else {
                j++;
                if (j >= n) break;
            }
        }
        return ans;
    }
};
```
### 滑动窗口思想（自编）
```c++
class Solution {
public:
    int numberOfSubstrings(string s) {
        if(s.size()<3){return 0;}
        int i=0;int j=2;int ans=0;
        vector<int> cnt(3,0);
        cnt[s[0]-'a']++;cnt[s[1]-'a']++;cnt[s[2]-'a']++;
        while(i<=s.size()-3){
            if(cnt[0]>0&&cnt[1]>0&&cnt[2]>0){
                ans+=s.size()-j;
                cnt[s[i]-'a']--;
                i++;
            }else{
                j++;
                if(j>=s.size()){break;}
                cnt[s[j]-'a']++;
            }
        }
        return ans;
    }
};
```
利用cnt记录状态，并根据区间的改变而调整cnt的状态
### 滑动窗口思想（更简洁）
```c++
class Solution {
public:
    int numberOfSubstrings(string s) {
        int n = s.size();
        int cnt[3] = {0};
        int i = 0, ans = 0;

        for (int j = 0; j < n; ++j) {
            cnt[s[j] - 'a']++;

            while (cnt[0] > 0 && cnt[1] > 0 && cnt[2] > 0) {
                ans += n - j;
                cnt[s[i] - 'a']--;
                i++;
            }
        }

        return ans;
    }
};
```
可以先不必关心前三个字符状态，在后面for循环中再去逐步更新，代码更简洁，行为更统一

## 题目三 替换后的最长重复字符
题干：给你一个字符串 s 和一个整数 k 。你可以选择字符串中的任一字符，并将其更改为任何其他大写英文字符。该操作最多可执行 k 次。在执行上述操作后，返回 包含相同字母的最长子字符串的长度。
```c++
class Solution {
public:
    int characterReplacement(string s, int k) {
        int i=0;
        int maxCount=0;
        int ans=0;
        vector<int> temp(26);
        for(int j=0;j<s.size();++j){
            temp[s[j]-'A']++;
            maxCount=max(temp[s[j]-'A'],maxCount);
            if(j-i+1>maxCount+k){
                temp[s[i]-'A']--;
                i++;
            }
            ans=max(ans,j-i+1);
        }
        return ans;
    }
};
```
当左右节点之间的距离大于其中最多字母的数量与操作次数之和时，表明操作次数已经不够使字符串满足题目定义形式了，即需要对节点进行调整了

# 滑动窗口总结
## 什么是滑动窗口
滑动窗口本质上是一种在数组或字符串这类**线性序列**上，**维护一个连续区间** [left, right] 的算法思想。这个区间就叫“窗口”。所谓“滑动”，就是指窗口的**左右边界都只朝一个方向移动**，从而在遍历过程中动态维护某种区间信息，而**不是每次都重新计算整个区间。**

## 滑动窗口的本质
当我们枚举很多“连续子数组 / 子串”时，如果**相邻两个区间之间只差一个或少数几个元素，就没有必要每次重新统计整个区间**，而可以利用上一个区间的信息增量更新。  
例如，从区间 [l, r] 变成 [l, r+1] 时，只多了一个新元素；从 [l, r] 变成 [l+1, r] 时，只少了一个旧元素。因此我们可以：右移右指针时，把新元素“加入窗口”、右移左指针时，把旧元素“移出窗口”、这就是滑动窗口最核心的“增量维护”思想。
我们经常会用两层循环这种结构去扫描数组，时间复杂度通常是O(n^2)，而滑动窗口就是为了降低这种结构的时间复杂度，即暴力做法通常是：1.枚举所有区间；2.对每个区间**重新统计**和、最大值、字符种类数、频次等信息。滑动窗口的做法是：仍然在考察区间，但**不重新扫描整个区间**，而是随着边界移动，**实时维护**窗口状态

## 滑动窗口适用条件
1. 问题对象必须是“连续空间”

也就是题目问的是：连续子数组、连续子串、某一段区间，如果题目问的是“任意子序列”“任意选若干个元素”，那通常不是滑动窗口。

2. 窗口状态可以被动态维护

也就是说，当窗口右扩或左缩时，能够快速更新区间信息，而不是必须重算整段区间。常见可维护的信息有：区间和、区间内每个字符/元素出现次数、区间内不同元素个数、某种约束是否满足

3. 条件具有单调性

指当窗口扩大或缩小时，某个性质会朝一个方向变化，从而使双指针能够只向前走、不回头。比如：窗口越大，区间和通常越大（元素非负）、窗口越大，包含的字符种类数不会减少、窗口越大，需要替换/删除的代价可能变大

## 滑动窗口的基本操作
1. 扩张

右指针右移，把新元素纳入窗口。作用是：让窗口更大、尝试让窗口满足某个条件
何时扩？窗口还不满足要求

2. 收缩

左指针右移，把旧元素移出窗口。作用是：让窗口更小、维持窗口合法、在满足条件的情况下寻找更优答案
何时缩？窗口已经满足要求：若要求最长：在合法前提下尽量保留更大窗口，所以通常只在非法时缩；若要求最短：在合法前提下尽量压缩，所以一满足就缩；若要求计数：要看能否从当前合法窗口推出一批答案，再缩左边继续统计

## 常见滑动窗口类型
1. 固定长度窗口

窗口大小固定，比如长度恰好为 k。  
典型问题：长度为 k 的子数组最大和；长度为 k 的子串中某种统计值的最优解  
做法是：先维护一个长度为 k 的窗口；每次右移一格；加入右边新元素，移出左边旧元素  

2. 可变长度窗口

窗口长度不固定，而是根据某个条件动态调整。  
典型问题：最长无重复子串；最小覆盖子串；和至少为 target 的最短子数组；最多替换 k 次后可得到的最长重复字符子串  
做法通常是：右指针不断扩张窗口；当窗口满足或不满足某条件时，再移动左指针调整

## 滑动窗口最常维护的状态
1. 区间和

    - 和大于等于某值
    - 固定长度最大和
    - 非负数组中的最短区间问题

2. 频次数组 / 哈希表

    常用于字符串和离散值问题：(vector<int> cnt(26, 0);unordered_map<char, int> mp;)
    - 统计字符出现次数
    - 判断是否重复
    - 维护窗口内不同元素个数

3. 不同元素个数

    - 用于“至多 k 种字符 / 元素”的问题。

4. 当前窗口是否满足某约束

    - 是否包含所有目标字符
    - 是否有重复元素
    - 替换代价是否不超过 k

## 最重要的三点
- 窗口的合法条件是什么
- 窗口中需要维护哪些状态
- 窗口满足或不满足条件时应当扩右还是缩左
